```r
library(ggplot2)
```

```r
df <- data.frame(
    x = c(0, 1, 2, 3, 4),
    y = c(1, 2, 3, 2, 1),
    gd = c('a', 'b', 'c', 'd', 'e'),
    s = c(1, 3, 5, 2, 4)
)
```

```r
ggplot(df, aes(x, y)) +
    geom_pointrange(aes(ymin = 0, ymax = y, size = gd, linewidth = gd), shape = 22, color = "black", fill = "red")
```

```r
ggplot(df, aes(x, y)) +
    geom_pointrange(aes(ymin = 0, ymax = y, size = s, linewidth = s), shape = 22, color = "black", fill = "red")
```

```r
ggplot(df, aes(x, y)) +
    geom_pointrange(aes(ymin = 0, ymax = y, size = s, linewidth = s), shape = 22, color = "black", fill = "red") +
    scale_size_identity() + scale_linewidth_identity()
```
