# Bitcoin Daily Trend Visualization (2024)

This notebook was inspired by a [Stack Overflow question](https://stackoverflow.com/questions/78801684/how-to-use-2-colors-on-same-line-lets-plot-in-kotlin).

```python
from datetime import timedelta

import pandas as pd

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
year = 2024
df = pd.read_csv("data/Bitstamp_BTCUSD_d.csv", parse_dates=["date"])
df = df[df["date"].dt.year == year].sort_values(by="date").reset_index(drop=True)
df = df.assign(trend=(df["close"].diff().fillna(0) > 0).astype(int))
print(df.shape)
df.head()
```

```python
ggplot(df) + \
    geom_line(aes("date", "close"), color="pen", size=1.2, tooltips='none') + \
    geom_line(aes("date", "close", color="trend"), size=.6, tooltips='none') + \
    scale_x_datetime(expand=[0]) + \
    scale_y_continuous(format="$,.4~g") + \
    scale_color_manual(name="Trend", values={0: "red", 1: "green"},
                       labels=["Downtrend", "Uptrend"]) + \
    labs(title="Bitcoin Daily Close Price with Up/Down Trend Coloring ({0})".format(year),
         x="", y="Close Price") + \
    ggsize(1000, 500) + \
    theme_light() + \
    flavor_darcula() + \
    theme(exponent_format="pow",
          plot_title=element_text(size=18, face='bold'),
          legend_position='bottom') + \
    ggtb()
```
