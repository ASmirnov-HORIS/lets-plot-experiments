```python
from math import sqrt, pi

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
def get_plot2(start):
    values = [i * sqrt(2) for i in range(start, start + 5)]
    binwidth = (max(values) - min(values)) / 3
    bin_centers = [min(values) + binwidth / 2 + i * binwidth for i in range(3)]
    breaks = [bin_centers[0] - binwidth / 2, bin_centers[1] + binwidth / 2, bin_centers[2] + binwidth / 2]
    return ggplot() + \
        geom_histogram(aes(x=values), breaks=breaks, color="white") + \
        geom_point(aes(x=values), color="red", size=5)

gggrid([
    get_plot2(start)
    for start in range(1, 10)
], ncol=3)
```
