## Example #1

```r
ggplot(diamonds, aes(price)) + geom_histogram(color = "black", fill = "white")
```

```r
price_bins <- quantile(diamonds$price, probs = seq(0, 1, length = 11))
ggplot(diamonds, aes(price)) + geom_histogram(breaks = price_bins, color = "black", fill = "white")
```

## Example #3

```r
df <- data.frame(
    v = c(1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 4, 4)
)
```

```r
def_p <- ggplot(df) + geom_histogram(aes(x = v), binwidth = 1, color = "black", fill = "white")
def_p
```

```r
layer_data(def_p, 1)
```

```r
breaks_p <- ggplot(df) + geom_histogram(aes(x = v), breaks = c(0, 3, 4), color = "black", fill = "white")
breaks_p
```

```r
layer_data(breaks_p, 1)
```

```r
ggplot(df) + geom_histogram(aes(x = v), breaks = c(-1, 0, 3, 4, 5), color = "black", fill = "white")
```

```r
ggplot(df) + geom_histogram(aes(x = v), breaks = c(0, 3), color = "black", fill = "white")
```

```r
ggplot(df) + geom_histogram(aes(x = v), breaks = c(3, 4), color = "black", fill = "white")
```

## Example 4

```r
identity_df <- data.frame(
    x <- c(1.5, 3.5),
    xmin <- c(0, 3),
    xmax <- c(3, 4),
    y <- c(6, 6)
)
```

```r
ggplot(identity_df) + geom_histogram(aes(x, y, xmin = xmin, xmax = xmax), breaks = c(0, 3, 4), stat = "identity")
```
