# Check vjust

```r
library(ggplot2)
library(ggpubr)
```

## ggplot2

### Default

```r
ggplot() + geom_point(aes(x = 0, y = 0), size = 25) + geom_text(aes(x = 0, y = 0, label = "TEXT"), size = 25, color = "red")
```

### vjust = 1

```r
ggplot() + geom_point(aes(x = 0, y = 0), size = 25) + geom_text(aes(x = 0, y = 0, label = "TEXT"), size = 25, color = "red", vjust = 1)
```

### vjust = 2

```r
ggplot() + geom_point(aes(x = 0, y = 0), size = 25) + geom_text(aes(x = 0, y = 0, label = "TEXT"), size = 25, color = "red", vjust = 2)
```

## ggpubr

### Default

```r
data <- data.frame(
    group1 = c("4", "f", "4"),
    group2 = c("f", "r", "r"),
    p = c(0.01, 0.02, 0.04),
    y.position = c(10, 8, 6)
)

p <- ggplot(mpg, aes(drv, hwy)) + geom_boxplot()
p + stat_pvalue_manual(data, tip.length = -0.02)
```

### vjust = 3

```r
p + stat_pvalue_manual(data, tip.length = -0.02, vjust = 3)
```
