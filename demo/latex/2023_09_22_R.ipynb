```r
library(latex2exp)
library(ggplot2)
```

```r
get_data <- function (a = -3, b = 3, n = 3) {
    size <- (b - a) * 2^n + 1
    xs <- seq(from = a, to = b, length.out = size)
    ys <- xs^2
    data.frame(x = xs, y = ys)
}
```

```r
df <- get_data()
```

```r
ggplot(df, aes(x, y)) + geom_line() + ggtitle(TeX("ggplot2: $y = x^2$"))
```
