```python
import pandas as pd

from lets_plot import *
LetsPlot.setup_html()

import os; import sys; sys.path.append(os.path.join(sys.path[0], ".."))
from util import util as u
from util import font as ufont
from util import clustering_model as ucm # TODO
from util import transform_data as utd
```

```python
cw_windows_df = u.read_data("../data/to_compare/char_widths_windows.csv")
print("Char widths for Windows data shape: {0}".format(cw_windows_df.shape))
cw_windows_df.head()
```

```python
cw_macos_df = u.read_data("../data/to_compare/char_widths_macos.csv")
print("Char widths for MacOS data shape: {0}".format(cw_macos_df.shape))
cw_macos_df.head()
```

```python
common_columns = list(filter(lambda c: c != "width", cw_windows_df.columns))
cw_df = cw_windows_df.merge(cw_macos_df, left_on=common_columns, right_on=common_columns, suffixes=["_windows", "_macos"])
cw_df["width_ratio"] = cw_df.width_windows / cw_df.width_macos
cw_df.head()
```

```python
basic_cw_df = utd.filter_by_font(cw_df, filters=["family", "size", "face"])
print(basic_cw_df.shape)
basic_cw_df.head()
```

```python
ggplot(basic_cw_df) + \
    geom_hline(yintercept=1, color="red") + \
    geom_bin2d(aes("width_windows", "width_ratio", fill="alphabet"), binwidth=[1, .1]) + \
    coord_fixed(ratio=10)
```
