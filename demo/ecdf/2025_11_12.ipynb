# Explore More ECDF

```r
library(ggplot2)
```

```r
df <- data.frame(
  x = c(1, -1, 2, -2, 3, -3, 1, -1)
)
head(df)
```

```r
p1 <- ggplot() + stat_ecdf(aes(x = x), df)
layer_data(p1, 1)
```

```r
p1
```

```r
ggplot() + stat_ecdf(aes(x = x), df, geom = "line")
```

```r
ggplot() + stat_ecdf(aes(x = x), df, geom = "path")
```

```r
n = 150000
set.seed(42)
df <- data.frame(
  x = rnorm(n, 0, 3)
)
head(df, 5)
```

```r
ggplot() + stat_ecdf(aes(x = x), df)
```

```r
ggplot() + stat_ecdf(aes(x = x), df, geom = "path")
```
