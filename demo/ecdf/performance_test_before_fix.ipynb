```python
import numpy as np

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
def get_big_data(n, *, seed=42):
    np.random.seed(seed)
    return {
        "x": np.random.normal(size=n),
        "g": np.random.choice(["A", "B", "C"], size=n),
    }

def get_big_data_plot(n, *, seed=42):
    data = get_big_data(n, seed=seed)
    return ggplot(data) + \
        stat_ecdf(aes(x="x", color="g"), show_legend=False) + \
        theme_void()
```

```python
size_min = 5_000
size_max = 50_000
step = 2_500
data = {'dataset_size': [], 'avg_time': []}
for size in range(size_min, size_max + 1, step):
    print(f"size={size}")
    t = %timeit -r 10 -n 2 -o get_big_data_plot(size)._repr_html_()
    data['dataset_size'].append(size)
    data['avg_time'].append(t.average)
```

```python
ggplot(data, aes("dataset_size", "avg_time")) + \
    geom_line() + \
    geom_point() + \
    ggsize(1000, 600)
```
