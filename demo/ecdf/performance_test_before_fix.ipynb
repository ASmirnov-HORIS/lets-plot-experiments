```python
import numpy as np
import pandas as pd

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
fix_column_value = "before"
```

```python
def get_big_data(n, *, seed=42):
    np.random.seed(seed)
    return {
        "x": np.random.normal(size=n),
        "g": np.random.choice(["A", "B", "C"], size=n),
    }

def get_big_data_without_groups(n, *, seed=42):
    np.random.seed(seed)
    return {
        "x": np.random.normal(size=n),
    }

def get_big_data_with_repetitions(n, *, m=None, seed=42):
    np.random.seed(seed)
    if m is None:
        m = int(n / 100) + 1
    values = np.random.normal(size=m)
    return {
        "x": np.random.choice(values, size=n),
        "g": np.random.choice(["A", "B", "C"], size=n),
    }

def get_big_data_plot(data, *, seed=42):
    mapping = aes(x="x", color="g") if "g" in data.keys() else aes(x="x")
    return ggplot(data) + \
        stat_ecdf(mapping, show_legend=False) + \
        theme_void()
```

```python
gggrid([
    get_big_data_plot(get_big_data(5_000)),
    get_big_data_plot(get_big_data_without_groups(5_000)),
    get_big_data_plot(get_big_data_with_repetitions(5_000)),
])
```

```python
size_min = 1_000
size_max = 101_000
step = 25_000
data = {'dataset': [], 'dataset_size': [], 'avg_time': []}
for get_data, dataset_name in [
    (get_big_data, "with groups"),
    (get_big_data_without_groups, "without groups"),
    (get_big_data_with_repetitions, "with repetitions"),
]:
    for size in range(size_min, size_max + 1, step):
        print(f"size={size}, dataset {dataset_name}")
        t = %timeit -r 5 -n 2 -o get_big_data_plot(get_data(size))._repr_html_()
        data['dataset'].append(dataset_name)
        data['dataset_size'].append(size)
        data['avg_time'].append(t.average)
df = pd.DataFrame(data).assign(fix=fix_column_value)
df.to_csv('{0}_fix.csv'.format(fix_column_value), index=False)
df.head()
```
