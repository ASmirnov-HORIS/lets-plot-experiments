```python
import numpy as np
import pandas as pd

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
fix_column_value = "before"
```

```python
def get_big_data(n, *, seed=42):
    np.random.seed(seed)
    return {
        "x": np.random.normal(size=n),
        "g": np.random.choice(["A", "B", "C"], size=n),
    }

def get_big_data_plot(n, *, sampling=None, seed=42):
    data = get_big_data(n, seed=seed)
    return ggplot(data) + \
        stat_ecdf(aes(x="x", color="g"), sampling=sampling, show_legend=False) + \
        theme_void()
```

```python
get_big_data_plot(1_000)
```

```python
size_min = 1_000
size_max = 101_000
step = 25_000
data = {'dataset_size': [], 'avg_time': [], 'sampling': []}
for sampling in [None, "none"]:
    for size in range(size_min, size_max + 1, step):
        print(f"sampling={sampling}, size={size}")
        t = %timeit -r 5 -n 2 -o get_big_data_plot(size, sampling=sampling)._repr_html_()
        data['dataset_size'].append(size)
        data['avg_time'].append(t.average)
        data['sampling'].append(sampling)
df = pd.DataFrame(data).assign(fix=fix_column_value)
df.to_csv('{0}_fix.csv'.format(fix_column_value), index=False)
df
```
