# ECDF

```r
library(ggplot2)
```

```r
set.seed(42)
df <- data.frame(
  x = c(rnorm(20, 0, 3), rnorm(20, 0, 10)),
  g = gl(2, 20)
)
head(df)
```

### Default

```r
ggplot(df, aes(x)) + stat_ecdf()
```

### All Parameters

```r
ggplot(df, aes(x)) + stat_ecdf(aes(color = g))
```

```r
ggplot() + stat_ecdf(aes(x), data = df)
```

```r
ggplot(df, aes(x)) + stat_ecdf(geom = 'point')
```

```r
ggplot(df, aes(x)) + stat_ecdf(n = 10)
```

```r
ggplot(df, aes(x)) + stat_ecdf(pad = FALSE)
```
