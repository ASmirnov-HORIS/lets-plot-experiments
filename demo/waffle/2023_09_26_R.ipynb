# Waffle plot

Exploration of the `geom_waffle()` from the https://github.com/hrbrmstr/waffle.

```r
install.packages("plyr")
```

```r
devtools::install_git("https://git.rud.is/hrbrmstr/waffle.git")
```

```r
library(ggplot2)
library(waffle)
library(plyr)
```

```r
df <- data.frame(
    g <- c('a', 'b', 'c'),
    v <- c(12, 8, 3)
)
```

#### Default

```r
ggplot(df, aes(fill = g, values = v)) + geom_waffle(na.rm = FALSE)
```

#### `mapping` and `data` parameters

```r
ggplot() + geom_waffle(aes(fill = g, values = v), data = df, na.rm = FALSE)
```

#### `n_rows` parameter

```r
ggplot(df, aes(fill = g, values = v)) + geom_waffle(n_rows = 5, na.rm = FALSE)
```

#### `make_proportional` parameter

```r
ggplot(df, aes(fill = g, values = v)) + geom_waffle(make_proportional = TRUE, na.rm = FALSE)
```

#### `flip` parameter

```r
ggplot(df, aes(fill = g, values = v)) + geom_waffle(flip = TRUE, na.rm = FALSE)
```

#### `radius` parameter

```r
ggplot(df, aes(fill = g, values = v)) + geom_waffle(radius = grid::unit(0.25, "npc"), na.rm = FALSE)
```

#### Other parameters and layers

```r
ggplot(df, aes(fill = g, values = v)) + 
    geom_waffle(n_rows = 5, na.rm = FALSE, color = 'white', size = 2, show.legend = FALSE) +
    scale_fill_brewer(type = 'qual', palette = 'Accent') +
    coord_fixed(ratio = 0.75) +
    ggtitle("Waffle plot") +
    theme_void()
```
