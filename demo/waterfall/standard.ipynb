## Tests

### Boundary Value Analysis

```python
import numpy as np

class BVATest:
    def __init__(self, data, title, show=True):
        self.data = data
        self.title = title
        self.show = show

    def to_plot(self):
        return gggrid([
            waterfall_plot(self.data, 'x', 'y', calc_total=True) + \
                ggtitle(self.title, "calc_total=True"),
            waterfall_plot(self.data, 'x', 'y', calc_total=False) + \
                ggtitle(self.title, "calc_total=False"),
        ])

show_all = True
bva_tests = [
    BVATest(
        data=dict(
            x=[],
            y=[],
        ),
        title="Empty dataset",
        show=False,
    ),
    BVATest(
        data=dict(
            x=["A"],
            y=[1],
        ),
        title="One value dataset",
    ),
    BVATest(
        data=dict(
            x=["A"],
            y=[0],
        ),
        title="Zero values",
    ),
    BVATest(
        data=dict(
            x=["A"],
            y=[-1],
        ),
        title="Negative values",
    ),
    BVATest(
        data=dict(
            x=[1],
            y=[1],
        ),
        title="Numeric x",
    ),
    BVATest(
        data=dict(
            x=["A", "B", "C"],
            y=[1, float('inf'), float('-inf')],
        ),
        title="Inf values",
    ),
    BVATest(
        data=dict(
            x=["A", "B", None],
            y=[1, None, 1],
        ),
        title="None values",
    ),
    BVATest(
        data=dict(
            x=["A", "B", np.nan],
            y=[1, np.nan, 1],
        ),
        title="np.nan values",
    ),
    BVATest(
        data=dict(
            x=["A", "B"],
            y=[1, -1],
        ),
        title="Total is zero",
    ),
]

gggrid([
    t.to_plot()
    for t in filter(lambda t: show_all or t.show, bva_tests)
], ncol=1)
```
