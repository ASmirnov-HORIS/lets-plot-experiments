## 1. Scales

```python
import datetime as dt
```

```python
n = 5
data = {
    'x': [dt.datetime(2024, 1, 1) + dt.timedelta(days=d) for d in range(n)],
    'y': [1]*n,
}
```

```python
waterfall_plot(data, 'x', 'y', calc_total=False) + scale_x_datetime(format='%Y-%m-%d')
```

```python
waterfall_plot(data, 'x', 'y', calc_total=False) + scale_x_continuous(breaks=[0, 2, 4], labels=['0', '2', '4'])
```
