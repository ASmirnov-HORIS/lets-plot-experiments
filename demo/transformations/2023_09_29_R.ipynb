```r
library(ggplot2)
library(gridExtra)
library(repr)
```

```r
options(warn=-1, repr.plot.width = 16, repr.plot.height = 12)
```

```r
get_data <- function (n) {
    x <- seq(-n, n)
    exp <- exp(1)^x
    exp10 <- 10^x
    exp2 <- 2^x
    data.frame(x = x, exp = exp, exp10 = exp10, exp2 = exp2)
}
```

```r
df <- get_data(4)
tail(df)
```

```r
get_plot <- function (col, trans) {
    ggplot(df, aes(x = x, y = get(col))) +
        geom_point() +
        scale_y_continuous(trans = trans) +
        ggtitle(sprintf("col = '%s', trans = '%s'", col, trans))
}
```

```r
grid.arrange(
    get_plot('exp', 'log'), get_plot('exp10', 'log'), get_plot('exp2', 'log'),
    get_plot('exp', 'log10'), get_plot('exp10', 'log10'), get_plot('exp2', 'log10'),
    get_plot('exp', 'log2'), get_plot('exp10', 'log2'), get_plot('exp2', 'log2'),
ncol = 3)
```
