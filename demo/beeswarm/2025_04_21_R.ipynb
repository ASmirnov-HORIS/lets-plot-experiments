### Explore `method = "counts"`

```r
data <- data.frame(
    x = c('a', 'a', 'a', 'a', 'a', 'a', 'a', 'a'),
    y = c(1, 1, 2, 2, 1, 1, 1, 2),
    g = c('x', 'x', 'x', 'x', 'y', 'y', 'y', 'y')
)
```

```r
options(repr.plot.width = 16, repr.plot.height = 6)
```

```r
p1 <- ggplot(data, aes(x, y, fill = g)) +
    geom_violin(alpha = .25) +
    geom_sina(shape = 21, size = 5, jitter_y = FALSE, seed = 42) +
    ggtitle("Default method")
p2 <- ggplot(data, aes(x, y, fill = g)) +
    geom_violin(alpha = .25) +
    geom_sina(method = "counts", bins = 2, shape = 21, size = 5, jitter_y = FALSE, seed = 42) +
    ggtitle("method = \"counts\"")

plot_grid(
    p1, p2,
    ncol = 2
)
```

```r
pg <- ggplot_build(p2)
pg$data[1]
```
