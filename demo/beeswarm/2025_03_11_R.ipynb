# Sina plot and Beeswarm plot (violin point plot)

See:

- [Issue #1023](https://github.com/JetBrains/lets-plot/issues/1023) - Add `geom_beeswarm()`.
- [Issue #1298](https://github.com/JetBrains/lets-plot/issues/1298) - Feature request `geom_sina()`.
- [Sina plot](https://ggforce.data-imaginist.com/reference/geom_sina.html).
- [Beeswarm-style plots](https://github.com/eclarke/ggbeeswarm).

```r
library(ggplot2)
library(ggforce)
library(ggbeeswarm)
```

```r
p <- ggplot(iris, aes(Species, Sepal.Length))
```

## `geom_sina()` from ggforce

Fits the violin plot perfectly, has most of the same parameters.

```r
p + geom_violin() + geom_sina(color = 'red', seed = 42)
```

## `geom_beeswarm()` from ggbeeswarm

Here, points only fall into the same group if their y-coordinates are exactly the same, as in `position_stack()`. Unlike dotplot/ydotplot (LP) where there is a concept of bin.

```r
p + geom_violin(scale = 'width', width = .4) + geom_beeswarm(color = 'red')
```

## `geom_quasirandom()` from ggbeeswarm

Doesn't have all the same parameters as a `geom_violin()`, so can't be adapted to any violin plot. But it has its own set of methods for customising the shape of the plot. Essentially it is the same as `geom_beeswarm()`, but the points are additionally shifted by some random component, the value of which is determined by the `method`.

```r
set.seed(42)
```

```r
p + geom_violin(scale = 'width', width = .8) + geom_quasirandom(color = 'red')
```
