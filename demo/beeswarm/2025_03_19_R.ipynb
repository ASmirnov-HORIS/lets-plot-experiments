# Explore Position

```r
library(ggplot2)
library(ggforce)
```

```r
data <- data.frame(
    x = c('a', 'a', 'a', 'a', 'a', 'a'),
    y = c(1, 1, 2, 1, 1, 2),
    g = c('x', 'x', 'x', 'y', 'y', 'y')
)
```

```r
ggplot(data, aes(x, y, fill = g)) +
    geom_violin(alpha = .25) +
    geom_sina(shape = 21, size = 5)
```
