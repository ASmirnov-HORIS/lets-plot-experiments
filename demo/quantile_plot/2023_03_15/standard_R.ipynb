```r
library(ggplot2)
```

```r
mapping <- aes(color = after_stat(quantile))
quantiles <- c(0.01, 0.25, 0.5, 0.75, 0.99)
scale_layer <- scale_color_gradient(low = "red", high = "green")
```

```r
df <- data.frame(
    x = c(0, 0, 1, 1, 1, 2, 2, 3),
    y = c(0, 1, 0, 1, 2, 1, 2, 3)
)
df
```

```r
ggplot(df, aes(x, y)) +
    geom_point(size = 5) +
    geom_quantile(mapping, quantiles = quantiles, linewidth = 1) +
    scale_layer
```

```r
ggplot(iris, aes(Petal.Length, Petal.Width)) +
    geom_point(size = 3) +
    geom_quantile(mapping, quantiles = quantiles, linewidth = 1) +
    scale_layer
```
