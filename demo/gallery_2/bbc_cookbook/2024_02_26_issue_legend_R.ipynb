```r
library(ggplot2)
library(cowplot)
```

```r
options(repr.plot.width=16, repr.plot.height=12)
```

```r
p <- ggplot() +
    geom_point(aes(x = c(0, 1), y = c(0, 1), color = c("A", "B")))
```

```r
grid <- plot_grid(
    p + theme(legend.position = "right", legend.justification = "center") + ggtitle('legend.position = "right", legend.justification = "center" (default)'),
    p + theme(legend.position = "top", legend.justification = "center") + ggtitle('legend.position = "top", legend.justification = "center"'),
    p + theme(legend.position = "right", legend.justification = "top") + ggtitle('legend.position = "right", legend.justification = "top"'),
    p + theme(legend.position = "top", legend.justification = "top") + ggtitle('legend.position = "top", legend.justification = "top"'),
    p + theme(legend.position = "right", legend.justification = "right") + ggtitle('legend.position = "right", legend.justification = "right"'),
    p + theme(legend.position = "top", legend.justification = "right") + ggtitle('legend.position = "top", legend.justification = "right"'),
    ncol = 2
)

print(grid)
```
