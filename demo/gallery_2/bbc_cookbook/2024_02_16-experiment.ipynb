```python
import pandas as pd

from lets_plot import *
from lets_plot.mapping import as_discrete
```

```python
LetsPlot.setup_html()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/gapminder.csv")
df.head()
```

```python
def lets_plot_bbc_style(show_x_line=True):
    def get_element_text(title=False, subtitle=False):
        face = None
        size = 18
        margin = None
        if title:
            size = 28
            face = "bold"
        if subtitle:
            size = 22
            margin = [9, 0, 9, 0]
        return element_text(family="Helvetica", face=face, size=size, margin=margin)
    bbc_theme = theme(
        plot_title=get_element_text(title=True),
        plot_subtitle=get_element_text(subtitle=True),
        legend_position='top',
        legend_background='blank',
        legend_title='blank',
        legend_text=get_element_text(),
        axis_title='blank',
        axis_text=get_element_text(),
        axis_text_x=element_text(margin=[5, 10]),
        axis_ticks='blank',
        axis_line=element_line(size=2.2) if show_x_line else 'blank',
        axis_ontop_x=True,
        panel_grid_minor='blank',
        panel_grid_major_x='blank',
        panel_background='blank',
        strip_text=element_text(size=22, hjust=0)
    )
    if show_x_line:
        bbc_theme += geom_hline(yintercept=0, size=0, tooltips='none')
    return bbc_theme
```

## Make a line chart

```python
line_df = df[df.country == "Malawi"]

p = ggplot(line_df, aes('year', 'lifeExp')) + \
    geom_line(color='#1380A1', size=1, \
              tooltips=layer_tooltips().format("@year", "d")) + \
    scale_x_continuous(format='d') + \
    lets_plot_bbc_style() + \
    labs(title="Living longer", subtitle="Life expectancy in Malawi 1952-2007")
p
```
