```r
library(ggplot2)
library(cowplot)
```

```r
options(repr.plot.width=16, repr.plot.height=20)
```

```r
p <- ggplot() +
    geom_point(aes(x = 0, y = 0)) +
    ggtitle("Title", subtitle = "Subtitle") +
    theme(plot.caption = element_text(size = 18))
```

```r
grid <- plot_grid(
    p + labs(caption = "hjust = 0 (default); plot.title.position = 'panel' (default)"),
    p + theme(plot.title.position = "plot") + labs(caption = "hjust = 0 (default); plot.title.position = 'plot'"),
    p + theme(plot.title = element_text(hjust = 1)) + labs(caption = "hjust = 1; plot.title.position = 'panel' (default)"),
    p + theme(plot.title = element_text(hjust = 1), plot.title.position = "plot") + labs(caption = "hjust = 1; plot.title.position = 'plot'"),
    p + theme(plot.title = element_text(hjust = 0.5)) + labs(caption = "hjust = 0.5; plot.title.position = 'panel' (default)"),
    p + theme(plot.title = element_text(hjust = 0.5), plot.title.position = "plot") + labs(caption = "hjust = 0.5; plot.title.position = 'plot'"),
    p + theme(plot.title = element_text(margin = margin(l = 100))) + labs(caption = "margin = margin(l = 100); plot.title.position = 'panel' (default)"),
    p + theme(plot.title = element_text(margin = margin(l = 100)), plot.title.position = "plot") + labs(caption = "margin = margin(l = 100); plot.title.position = 'plot'"),
    ncol = 2
)

print(grid)
```
