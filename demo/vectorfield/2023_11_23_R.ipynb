```r
library(ggplot2)
```

```r
get_data <- function (n = 10, seed = 42) {
    set.seed(seed)

    result <- expand.grid(x = 1:n, y = 1:n)
    result$angle <- runif(n^2, 0, 2*pi)
    result$radius <- runif(n^2, 0, sqrt(.1 * result$x))
    result
}
```

```r
df <- get_data()
head(df)
```

```r
p1 <- ggplot(df, aes(x, y)) + geom_spoke(aes(angle = angle), radius = .5)
p1
```

```r
p2 <- p1 + geom_point()
p2
```

```r
p3 <- ggplot(df, aes(x, y)) + geom_point() + geom_spoke(aes(angle = angle, radius = radius))
p3
```

```r
p4 <- ggplot(df, aes(x, y)) + geom_spoke(aes(angle = angle), radius = .5, linewidth = 1)
p4
```

```r
p5 <- ggplot(df, aes(x, y)) + geom_spoke(aes(angle = angle, radius = radius, color = radius))
p5
```
