```python
import pandas as pd

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
commits_df = pd.read_csv("data/commits.csv", sep=";;", engine='python', parse_dates=["date"])
commits_df["deletions"] = -commits_df["deletions"]
commits_df["deletions_kt"] = -commits_df["deletions_kt"]
commits_df["deletions_py"] = -commits_df["deletions_py"]
print(commits_df.shape)
commits_df.head()
```

```python
releases_df = pd.read_csv("data/releases.csv", parse_dates=["date"])
print(releases_df.shape)
releases_df.head()
```

```python
ggplot() + \
    geom_histogram(aes(x="date", y="..count..", weight="additions_kt"), data=commits_df, alpha=.5, fill="red",
                   tooltips=layer_tooltips().title("Kotlin").line("additions count|@..count..").line("date|@date")) + \
    geom_histogram(aes(x="date", y="..count..", weight="additions_py"), data=commits_df, alpha=.5, fill="blue",
                   tooltips=layer_tooltips().title("Python").line("additions count|@..count..").line("date|@date")) + \
    geom_histogram(aes(x="date", y="..count..", weight="deletions_kt"), data=commits_df, alpha=.5, fill="red",
                   tooltips=layer_tooltips().title("Kotlin").line("deletions count|@..count..").line("date|@date")) + \
    geom_histogram(aes(x="date", y="..count..", weight="deletions_py"), data=commits_df, alpha=.5, fill="blue",
                   tooltips=layer_tooltips().title("Python").line("deletions count|@..count..").line("date|@date")) + \
    geom_point(aes(x="date", fill="kind"), y=20_000, data=releases_df, shape=21,
               tooltips=layer_tooltips(["version", "kind", "date"])) + \
    coord_polar() + \
    scale_x_datetime(expand=[0, 0]) + \
    scale_y_continuous(trans='symlog') + \
    ggsize(1000, 800) + \
    theme_void()
```
