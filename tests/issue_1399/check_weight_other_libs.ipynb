```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def get_dens_data(v):
    return {
        'v': [0, 1, v],
        'w': [1, 2, None],
    }
```

```python
def plot_kde(v):
    plt.figure(figsize=(8, 6))
    sns.kdeplot(data=get_dens_data(v), x='v', weights='w', fill=True)
    plt.xlim(0, v)
    plt.show()
```

```python
plot_kde(2)
```

```python
plot_kde(10)
```

```python
plot_kde(50)
```
