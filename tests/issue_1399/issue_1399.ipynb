```python
data = {
    'x': [0, 0, 1, 1],
    'y': [0, 1, 0, 1],
    'w': [2, 1, 1, None],
}

ggplot(data) + geom_density2d(aes('x', 'y', weight='w'))
```

```python
plots = []
for v in [1, 2, 25, 100]:
    v_data = {
        'x': [0, 0, 1, v],
        'y': [0, 1, 0, v],
        'w': [2, 1, 1, None],
    }
    plots.append(ggplot(v_data) + geom_density2d(aes('x', 'y', weight='w')) + ggtitle(f"Point ({v}, {v}) with weight NA"))
gggrid(plots, ncol=2)
```
