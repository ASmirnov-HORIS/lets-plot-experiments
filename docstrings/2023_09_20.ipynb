```python
import numpy as np
from lets_plot import *
LetsPlot.setup_html()
n = 100
np.random.seed(42)
x = np.random.uniform(size=n)
y = np.random.normal(size=n)
g = np.random.choice(["A", "B"], size=n)
ggplot({'x': x, 'y': y, 'g': g}, aes(x='x', y='y', fill='g')) + \
    stat_summary_bin(geom='crossbar', bins=6, fatten=5, position='dodge')
```
