```r
library(ggplot2)
```

```r
ggplot() +
    geom_point(aes(x = 0.0, y = 0.0), size=50, shape=21, stroke=1, color="green", fill="black") +
    geom_point(aes(x = 2.0, y = 0.0), size=50) +
    geom_point(aes(x = 4.0, y = 0.0), size=50, shape=21, stroke=50, color="red", fill="black", alpha=0.5) +
    xlim(-1, 5)
```

```r
ggplot() +
    geom_point(aes(x = 0.0, y = 0.0), size=75, shape=21, stroke=75, color="blue", fill="red", alpha=.25) +
    geom_point(aes(x = 1.0, y = 0.0), size=75, shape=21, stroke=75, color="blue", fill="red", alpha=.25) +
    xlim(-2, 3)
```
