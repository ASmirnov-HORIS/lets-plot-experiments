```r
library(ggplot2)
```

```r
df <- data.frame(
    x = c(0, 0, 1, 1, 0, 0, 1, 1),
    f = c('A', 'B', 'B', 'A', 'B', 'A', 'A', 'B'),
    g = c('X', 'X', 'X', 'X', 'Y', 'Y', 'Y', 'Y')
)
```

```r
ggplot(df) +
    geom_bar(aes(x = x, fill = f)) +
    facet_grid(cols = vars(g))
```

```r
ggplot(df) +
    geom_bar(aes(x = x, fill = f), position = 'dodge') +
    facet_grid(cols = vars(g))
```
