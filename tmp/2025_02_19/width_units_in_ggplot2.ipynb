```r
library(ggplot2)
```

## Bar

```r
ggplot(data.frame(x = c('a', 'a', 'b')), aes(x = x)) +
    geom_bar(width = .9) +
    ggtitle("width = 0.9")
```

```r
ggplot(data.frame(x = c(0, 0, 10)), aes(x = x)) +
    geom_bar(width = .9) +
    ggtitle("width = 0.9")
```

```r
ggplot(data.frame(x = c(0, 0, 10)), aes(x = x)) +
    geom_bar(width = 9) +
    ggtitle("width = 9")
```

## Errorbar

```r
df1 <- data.frame(
    x = c('a', 'b'),
    ymin = c(1, 1),
    ymax = c(2, 3)
)
df2 <- data.frame(
    x = c(0, 10),
    ymin = c(1, 1),
    ymax = c(2, 3)
)
```

```r
ggplot(df1) +
    geom_errorbar(aes(x = x, ymin = ymin, ymax = ymax), width = .9) +
    ggtitle("width = 0.9")
```

```r
ggplot(df2) +
    geom_errorbar(aes(x = x, ymin = ymin, ymax = ymax), width = .9) +
    ggtitle("width = 0.9")
```

```r
ggplot(df2) +
    geom_errorbar(aes(x = x, ymin = ymin, ymax = ymax), width = 9) +
    ggtitle("width = 9")
```

## Tile

```r
df <- data.frame(
    x = c(-5, -5, 5),
    y = c(-5, 5, -5)
)
```

```r
ggplot(df, aes(x, y)) +
    geom_tile(linewidth = 1, color = "white") +
    ggtitle("default width")
```

```r
ggplot(df, aes(x, y)) +
    geom_tile(linewidth = 1, color = "white", width = .9) +
    ggtitle("width = 0.9")
```

```r
ggplot(df, aes(x, y)) +
    geom_tile(linewidth = 1, color = "white", width = 9) +
    ggtitle("width = 9")
```
