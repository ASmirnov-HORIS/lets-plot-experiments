```python
from lets_plot import *
```

```python
LetsPlot.setup_html(no_js=True, offline=True)
```

```python
data = {'xmin': [-3, -1], 'xmax': [1, 3], 'ymin': [-.5, -.5], 'ymax': [.5, .5], 'g': ['a', 'b']}
```

```python
def rect_plot(*, alpha=1):
    return ggplot(data) + \
        geom_rect(aes(xmin='xmin', xmax='xmax', ymin='ymin', ymax='ymax', color='g', fill='g'), alpha=alpha, size=10) + \
        ggtitle("geom_rect()", "alpha={0}".format(alpha))

def band_plot(*, alpha=1):
    return ggplot(data) + \
        geom_band(aes(xmin='xmin', xmax='xmax', color='g', fill='g'), alpha=alpha, size=10) + \
        ggtitle("geom_band()", "alpha={0}".format(alpha))
```

```python
gggrid([
    rect_plot(),
    rect_plot(alpha=.25),
    band_plot(),
    band_plot(alpha=.25),
], ncol=2)
```
