```r
library(ggplot2)
```

```r
df <- data.frame(x = c(0, 1), y = c(0, 0), angle = c(0.52, 1.57))
df
```

```r
ggplot(df, aes(x, y)) + geom_spoke(aes(angle = angle), radius = 0.5) + coord_fixed()
```

```r
ggplot(df, aes(x, y)) + geom_spoke(aes(angle = angle), radius = 0.5) + coord_flip() + theme(aspect.ratio = 2)
```
