```r
library(ggplot2)
library(repr)
library(cowplot)
```

```r
options(repr.plot.width=16, repr.plot.height=6)
```

```r
df1 <- data.frame(
    x = c(0, 0, 0),
    g = c("a", "a", "b")
)
```

```r
p1 <- ggplot(df1, aes(x, fill = g)) + geom_bar()
plot_grid(
    p1,
    p1 + ylim(0, 1)
)
```

```r
ggplot_build(p1)$data
```

```r
p2 <- ggplot(df1, aes(x, fill = g)) + geom_bar(position = 'fill')

plot_grid(
    p2,
    p2 + ylim(0, 1)
)
```

```r
ggplot_build(p2)$data
```

```r
df2 <- data.frame(
    x = c(0, 0, 0),
    y = c(-1, 0, 1),
    t = c("a", "b", "c")
)
```

```r
p3 <- ggplot(df2, aes(x, y)) + geom_text(aes(label = t), position = position_nudge(y = 1))

plot_grid(
    p3,
    p3 + ylim(-.5, 1.5)
)
```

```r
ggplot_build(p3)$data
```
