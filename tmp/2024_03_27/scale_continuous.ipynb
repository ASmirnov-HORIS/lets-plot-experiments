```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
data = {
    'x': [0, 0, 0, 0],
    'y': [1, 2, 3, 4],
    'val': [1, 2, 3, 50],
}
```

```python
point_plot = ggplot(data, aes('x', 'y')) + geom_point(aes(size='val'))

gggrid([
    point_plot + ggtitle("scale: default\nx limits: default"),
    point_plot + xlim(-10, 10) + ggtitle("scale: default\nx limits: -10<=x<=10"),
    point_plot + scale_size_identity() + ggtitle("scale: identity\nx limits: default"),
    point_plot + scale_size_identity() + xlim(-10, 10) + ggtitle("scale: identity\nx limits: -10<=x<=10"),
    point_plot + scale_continuous('size', name="radius", scale='size_area', range=[5, 10]) + ggtitle("scale: continuous\nx limits: default"),
    point_plot + scale_continuous('size', name="radius", scale='size_area', range=[5, 10]) + xlim(-10, 10) + ggtitle("scale: continuous\nx limits: -10<=x<=10"),
], ncol=2)
```

```python
spoke_plot = ggplot(data, aes('x', 'y')) + geom_spoke(aes(radius='val'))

gggrid([
    spoke_plot + ggtitle("scale: default\nx limits: default"),
    spoke_plot + xlim(-100, 100) + ggtitle("scale: default\nx limits: -100<=x<=100"),
    spoke_plot + scale_continuous('radius', name="radius", scale='size_area', range=[5, 10]) + ggtitle("scale: continuous\nx limits: default"),
    spoke_plot + scale_continuous('radius', name="radius", scale='size_area', range=[5, 10]) + xlim(-100, 100) + ggtitle("scale: continuous\nx limits: -100<=x<=100"),
], ncol=2)
```
