```python
data = {
    'x': [1, 2, 3, 4],
    'y': [0, 0, 0, 0],
    'size': [0, 2, 4, 32]
}
```

```python
p = ggplot(data, aes('x', 'y')) + geom_point(aes(size='size'))

gggrid([
    p + ggtitle("Default"),
    p + scale_size_identity() + ggtitle("scale_size_identity()"),
    p + scale_size() + ggtitle("scale_size()"),
    p + scale_size_area() + ggtitle("scale_size_area()"),
], ncol=2) + ggsize(1200, 900)
```

```python
import numpy as np

x = [1, 2, 3, 4]
y = [0, 0, 0, 0]
r = [0, 2, 4, 32]
a = [np.sqrt(v / np.pi) for v in r]
data = dict(x=x, y=y, r=r, a=a)
```

```python
g = ggplot(data, aes('x', 'y'))

gggrid([
    g + geom_point(aes(size='r')) + scale_size(range=[1, 20]),
    g + geom_point(aes(size='a')) + scale_size(range=[1, 20]),
]) + ggsize(1200, 450)
```
