```r
library(ggplot2)
library(repr)
library(cowplot)
```

```r
df <- data.frame(
    x = c(1, 2, 3, 4),
    y = c(0, 0, 0, 0),
    size = c(0, 2, 4, 32)
)
```

```r
options(repr.plot.width=36, repr.plot.height=18)
```

```r
p <- ggplot(df, aes(x, y)) + geom_point(aes(size = size))

plot_grid(
    p + ggtitle("Default"),
    p + scale_size_identity() + ggtitle("scale_size_identity()"),
    p + scale_radius() + ggtitle("scale_radius()"),
    p + scale_size() + ggtitle("scale_size()"),
    p + scale_size_area() + ggtitle("scale_size_area()"),
    ncol = 3
)
```

```r
plot_grid(
    p + scale_size(range = c(1, 20)) + ggtitle("scale_size()"),
    p + scale_radius(range = c(1, 20)) + ggtitle("scale_radius()"),
    ncol = 2
)
```
