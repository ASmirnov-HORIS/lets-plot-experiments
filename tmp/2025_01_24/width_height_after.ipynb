```python
LetsPlot.setup_html()
```

## `stat='identity'`

```python
def get_identity_plot(hexwidth, hexheight, width, height):
    dx = width * hexwidth
    dy = height * hexheight
    return ggplot() + \
        geom_hex(x=0, y=0,
                 hexwidth=hexwidth, hexheight=hexheight,
                 width=width, height=height,
                 stat='identity') + \
        coord_fixed(ratio=1, xlim=[-15, 15], ylim=[-15, 15]) + \
        ggtitle("hexwidth={hw}, hexheight={hh}\nwidth={w}, height={h}".format(
            hw=hexwidth,
            hh=hexheight,
            w=width,
            h=height,
        ))

gggrid([
    get_identity_plot(10, 10, 1, 1),
    get_identity_plot(20, 20, 1, 1),
    get_identity_plot(10, 10, 2, 2),
])
```

## `stat='binhex'`

```python
def get_binhex_plot(binwidth, width=None, height=None):
    data = {
        'x': [0, 6, 12, 18],
        'y': [0, 12, 0, 12],
    }
    return ggplot(data, aes('x', 'y')) + \
        geom_hex(binwidth=binwidth, width=width, height=height,
                 size=1, color="white") + \
        geom_point(color="red") + \
        coord_fixed(ratio=1, xlim=[-12, 30], ylim=[-12, 24]) + \
        ggtitle("binwidth=[{bw}, {bh}]\nwidth={w}, height={h}".format(
            bw=binwidth[0],
            bh=binwidth[1],
            w=width,
            h=height,
        ))

gggrid([
    get_binhex_plot([12, 12]),
    get_binhex_plot([12, 12], width=1, height=1),
    get_binhex_plot([12, 12], width=.5, height=.5),
    get_binhex_plot([12, 12], width=1.5, height=1.5),
    get_binhex_plot([6, 6], width=1, height=1),
    get_binhex_plot([18, 18], width=1, height=1),
], ncol=2)
```
