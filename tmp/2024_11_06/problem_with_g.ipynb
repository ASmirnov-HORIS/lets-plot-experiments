```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
def plot(f, dmin, dmax):
    value = 3.14
    values = [value * 10**d for d in range(dmin, dmax + 1)]

    return ggplot({'v': values}, aes(y='v')) + \
        geom_label(aes(label='v'), x=0, label_format=f) + \
        scale_y_log10(limits=[10**(dmin - 1), 10**(dmax + 2)])
```

#### Too small values

```python
plot("~g", -19, -16)
```

```python
plot("e", -19, -16)
```

#### Too big values

```python
plot("~g", 16, 19)
```

```python
plot("e", 16, 19)
```

```python
plot("~g", 16, 18)
```

```python
plot("", 16, 19)
```
