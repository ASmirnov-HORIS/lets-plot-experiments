```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
def plot(dmin, dmax):
    value = 3.14
    values = [value * 10**d for d in range(dmin, dmax + 1)]

    return ggplot({'v': values}, aes(y='v')) + \
        geom_label(aes(label='v'), x=0, label_format="~g") + \
        scale_y_log10(limits=[10**(dmin - 1), 10**(dmax + 2)])
```

```python
plot(-20, -16)
```

```python
plot(16, 18)
```

```python
plot(16, 19)
```
