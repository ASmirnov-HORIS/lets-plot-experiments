```r
iris_violin_plot <- ggplot(iris, aes(x = Species, y = Sepal.Length)) +
    geom_violin(trim = FALSE)
iris_violin_plot
```

```r
test_df <- data.frame(
    x = c("a", "a", "a", "b", "b", "b"),
    y = c(0, 0, 2, 1, 1, 3)
)
test_df
```

```r
test1_plot <- ggplot(test_df, aes(x = x, y = y)) +
    geom_violin(trim = FALSE)
test1_plot
```

```r
test1_plot_df <- as.data.frame(ggplot_build(test1_plot)$data)
head(test1_plot_df, 5)
```
