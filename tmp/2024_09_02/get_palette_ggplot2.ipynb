```r
install.packages("dplyr")
```

```r
library(dplyr)
library(ggplot2)
```

```r
df1 <- data.frame(
    x = c(1, 1, 2, 2),
    y = c(0, 1, 0, 1),
    g = c("A", "B", "C", "D"),
    v = c(0, 1, 2, 3)
)
filtered_df1 <- df1 %>% filter(v > 0)
```

## Default

```r
ggplot(df1, aes(x, y, color = g)) + geom_point(size=10)
```

```r
ggplot(filtered_df1, aes(x, y, color = g)) + geom_point(size=10)
```

## With custom values

```r
color <- scale_color_brewer(type = "qual", palette = "Set2")$palette(4)
color
```

```r
df2 <- df1
df2$color <- color
filtered_df2 <- df2 %>% filter(v > 0)
```

```r
ggplot(df2, aes(x, y, color = color)) + geom_point(size=10) + scale_color_identity()
```

```r
ggplot(filtered_df2, aes(x, y, color = color)) + geom_point(size=10) + scale_color_identity()
```
