```python
from lets_plot import *
LetsPlot.setup_html(offline=True, no_js=True)
```

```python
data = {
    'x': [0, 1],
    'y': [1, 2],
    'g': ['a', 'b'],
}
ggplot(data) + \
    geom_point(aes('x', 'y', color='g'), \
               position=position_dodge(width=.4), size=10)
```

```python
data = {
    'x': [0, 1, 1],
    'y': [1, 2, 3],
    'g': ['a', 'b', 'b'],
}
ggplot(data) + \
    geom_point(aes('x', 'y', color='g'), \
               position=position_dodge(width=.4), size=10)
```
