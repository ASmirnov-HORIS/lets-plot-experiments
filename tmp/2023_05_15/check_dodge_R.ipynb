```r
library(ggplot2)
```

```r
df <- data.frame(
    x = c(0, 1),
    y = c(1, 2),
    g = c('a', 'b')
)
ggplot(df) +
    geom_point(aes(x, y, color = g),
               position = position_dodge(width = .4), size = 10)
```

```r
df <- data.frame(
    x = c(0, 1, 1),
    y = c(1, 2, 3),
    g = c('a', 'b', 'b')
)
ggplot(df) +
    geom_point(aes(x, y, color = g),
               position = position_dodge(width = .4), size = 10)
```
