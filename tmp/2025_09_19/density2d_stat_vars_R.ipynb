```r
library(ggplot2)
library(cowplot)
```

```r
options(repr.plot.width = 14, repr.plot.height = 10)
```

```r
df <- data.frame(
    x = c(0, 0, 1),
    y = c(0, 1, 0)
)
```

```r
get_plot_grid <- function (g, n = 50) {
    plot_grid(
        g + geom_density2d(aes(color = after_stat(level))) + geom_point(size = 2, shape = 21, fill = "yellow"),
        g + geom_point(aes(color = after_stat(density)), stat = 'density2d', contour = FALSE, n = n),
        g + geom_point(aes(color = after_stat(count)), stat = 'density2d', contour = FALSE, n = n),
        g + geom_point(aes(color = after_stat(ndensity)), stat = 'density2d', contour = FALSE, n = n),
        ncol = 2
    )
}
```

```r
get_plot_grid(ggplot(df, aes(x, y)))
```
