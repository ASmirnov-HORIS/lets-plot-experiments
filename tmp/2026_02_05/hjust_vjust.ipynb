```r
library(ggplot2)
library("cowplot")
```

```r
get_plot <- function (hjust = 0.5, vjust = 0.5) {
    ggplot() +
        geom_point(aes(x = 0, y = 0), size = 30) +
        geom_text(aes(x = 0, y = 0, label = "TEXT"), size = 30, color = "red", hjust = hjust, vjust = vjust) +
        ggtitle(sprintf("hjust = %s, vjust = %s", hjust, vjust))
}
```

```r
options(repr.plot.width=16, repr.plot.height=12)
```

```r
plot_grid(
    get_plot(hjust = 0, vjust = 0),
    get_plot(hjust = 0, vjust = 0.5),
    get_plot(hjust = 0, vjust = 1),
    get_plot(hjust = 0.5, vjust = 0),
    get_plot(hjust = 0.5, vjust = 0.5),
    get_plot(hjust = 0.5, vjust = 1),
    get_plot(hjust = 1, vjust = 0),
    get_plot(hjust = 1, vjust = 0.5),
    get_plot(hjust = 1, vjust = 1),
    ncol = 3
)
```

```r
plot_grid(
    get_plot(hjust = 0.25, vjust = 0.25),
    get_plot(hjust = 0.25, vjust = 0.5),
    get_plot(hjust = 0.25, vjust = 0.75),
    get_plot(hjust = 0.5, vjust = 0.25),
    get_plot(hjust = 0.5, vjust = 0.5),
    get_plot(hjust = 0.5, vjust = 0.75),
    get_plot(hjust = 0.75, vjust = 0.25),
    get_plot(hjust = 0.75, vjust = 0.5),
    get_plot(hjust = 0.75, vjust = 0.75),
    ncol = 3
)
```

```r
plot_grid(
    get_plot(hjust = -0.5, vjust = -0.5),
    get_plot(hjust = -0.5, vjust = 0.5),
    get_plot(hjust = -0.5, vjust = 1.5),
    get_plot(hjust = 0.5, vjust = -0.5),
    get_plot(hjust = 0.5, vjust = 0.5),
    get_plot(hjust = 0.5, vjust = 1.5),
    get_plot(hjust = 1.5, vjust = -0.5),
    get_plot(hjust = 1.5, vjust = 0.5),
    get_plot(hjust = 1.5, vjust = 1.5),
    ncol = 3
)
```

```r
plot_grid(
    get_plot(hjust = "left", vjust = "bottom"),
    get_plot(hjust = "left", vjust = "center"),
    get_plot(hjust = "left", vjust = "top"),
    get_plot(hjust = "middle", vjust = "bottom"),
    get_plot(hjust = "middle", vjust = "center"),
    get_plot(hjust = "middle", vjust = "top"),
    get_plot(hjust = "right", vjust = "bottom"),
    get_plot(hjust = "right", vjust = "center"),
    get_plot(hjust = "right", vjust = "top"),
    ncol = 3
)
```

```r
plot_grid(
    get_plot() + xlim(0, 2) + ylim(0, 2),
    get_plot(hjust = "inward") + xlim(0, 2) + ylim(0, 2),
    get_plot(vjust = "inward") + xlim(0, 2) + ylim(0, 2),
    get_plot(hjust = "inward", vjust = "inward") + xlim(0, 2) + ylim(0, 2),
    get_plot(hjust = "outward") + xlim(0, 2) + ylim(0, 2),
    get_plot(vjust = "outward") + xlim(0, 2) + ylim(0, 2),
    get_plot(hjust = "outward", vjust = "outward") + xlim(0, 2) + ylim(0, 2),
    get_plot(hjust = "inward", vjust = "outward") + xlim(0, 2) + ylim(0, 2),
    get_plot(hjust = "outward", vjust = "inward") + xlim(0, 2) + ylim(0, 2),
    ncol = 3
)
```
