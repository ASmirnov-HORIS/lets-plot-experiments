```python
def dump_plot(plot, display=None):
    import json

    try:
        import clipboard
    except:
        clipboard = None
        
    from lets_plot._type_utils import standardize_dict
    
    plot_dict = standardize_dict(plot.as_dict())
    plot_json = json.dumps(plot_dict, indent=2)
    
    if clipboard:
        clipboard.copy('')
        clipboard.copy(str(plot_json))
    else:
        if display is None:
            display = True
    
    if display:
        print(plot_json)

    return plot

p = ggplot() + \
    geom_boxplot(aes(
        x=["x", "x", "x", "x", "x", "x", "x", "x"],
        y=[1, 2, 3, 4, 3, 4, 5, 6]
    ), linetype='dotted', size=14, fatten=0)

dump_plot(p)
```
