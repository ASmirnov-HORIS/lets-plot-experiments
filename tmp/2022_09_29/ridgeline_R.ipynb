```r
library(ggplot2)
library(ggridges)
```

```r
data <- data.frame(
    x = c(1, 2, 3, 1, 2, 3),
    y = c(1, 1, 1, 3, 3, 3),
    height = c(0.5, 1, 0.5, 0.5, 1, 0.5)
)
```

### `geom_ridgeline()` default

```r
ggplot(data) +
    geom_ridgeline(aes(x, y, height = height, group = y))
```

### `geom_ridgeline()` with `scale = 2`

```r
ggplot(data) +
    geom_ridgeline(aes(x, y, height = height, group = y), scale = 2)
```

### `geom_density_ridges()` with `stat = 'identity'`

```r
ggplot(data) +
    geom_density_ridges(aes(x, y, height = height, group = y), stat = 'identity')
```

### `geom_density_ridges()` with `stat = 'identity'` and `scale = 1`

```r
ggplot(data) +
    geom_density_ridges(aes(x, y, height = height, group = y), scale = 1, stat = 'identity')
```
