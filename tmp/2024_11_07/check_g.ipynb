```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
def plot_by_format(format_pattern):
    def plot(type):
        value = 123.45678901
        dmin, dmax = -4, 2
        label_format = None if type is None else format_pattern.format(type)
        title = "Default formatting" if type is None else "format=\"{0}\"".format(label_format)
        subtitle = "" if type is None else "type='{0}'".format(type)
        values = [value * 10**d for d in range(dmin, dmax + 1)]
        return ggplot({'v': values}, aes(y='v')) + \
            geom_label(aes(label='v'), label_format=label_format) + \
            scale_y_log10(limits=[value * 10**(dmin - 1), value * 10**(dmax + 1)]) + \
            ggtitle(title, subtitle)
    return plot

plot1 = plot_by_format(" >-,.6~{0}")
plot2 = plot_by_format(" >-,.2~{0}")
plot3 = plot_by_format("~{0}")
```

### With specified precision

```python
gggrid([
    plot1(None),
    plot1("f"),
    plot1("g"),
])
```

```python
gggrid([
    plot2(None),
    plot2("f"),
    plot2("g"),
])
```

### With default precision

```python
gggrid([
    plot3(None),
    plot3("f"),
    plot3("g"),
])
```
