## Problem \#1: NullPointerException when NaN values in dataset

```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")
df.head()
```

```python
# Original example
ggplot(df, aes("bill_length_mm", "bill_depth_mm")) + geom_density2d()
```

```python
# Minimal example
data = {
    'x': [0, 1, None],
    'y': [0, 1, None],
}
ggplot(data, aes('x', 'y')) + geom_density2d()
```

## Problem \#2: How to find out good value for the `binwidth` parameter?

```python
p = ggplot(df.dropna(), aes("bill_length_mm", "bill_depth_mm", color="species")) + coord_fixed(ratio=2)
p1 = p + geom_density2d() + ggtitle("Default")
p2 = p + geom_density2d(binwidth=1) + ggtitle("binwidth == 1")
p3 = p + geom_density2d(binwidth=.5) + ggtitle("binwidth == .5")
p4 = p + geom_density2d(binwidth=.1) + ggtitle("binwidth == .1")
p5 = p + geom_density2d(binwidth=.05) + ggtitle("binwidth == .05")
p6 = p + geom_density2d(binwidth=.01) + ggtitle("binwidth == .01")

w, h = 500, 375
bunch = GGBunch()
bunch.add_plot(p1, 0, 0, w, h)
bunch.add_plot(p2, w, 0, w, h)
bunch.add_plot(p3, 0, h, w, h)
bunch.add_plot(p4, w, h, w, h)
bunch.add_plot(p5, 0, 2 * h, w, h)
bunch.add_plot(p6, w, 2 * h, w, h)
bunch.show()
```
