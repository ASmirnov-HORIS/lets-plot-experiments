```python
data = {
    'val': [1, 2, 3],
    'min': [-2, -1, -2],
    'max': [1, 2, 0],
}

gggrid([
    ggplot(data) + geom_ribbon(aes(x='val', ymin='min', ymax='max')) + ggtitle("Correct tooltips"),
    ggplot(data) + geom_ribbon(aes(y='val', xmin='min', xmax='max')) + ggtitle("Wrong tooltips"),
])
```
