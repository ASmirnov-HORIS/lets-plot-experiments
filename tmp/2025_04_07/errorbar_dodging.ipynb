```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
data = {
    'xcat': ['x', 'x', 'y', 'y'],
    'xval': [0, 0, 1, 1],
    'ymin': [-2, -1, -2, -1],
    'ymax': [1, 2, 1, 2],
    'g': ['a', 'b', 'a', 'b'],
}

gggrid([
    ggplot(data) + geom_errorbar(aes(x='xcat', ymin='ymin', ymax='ymax', color='g')) + \
        ggtitle("x: discrete\nposition: default"),
    ggplot(data) + geom_errorbar(aes(x='xcat', ymin='ymin', ymax='ymax', color='g'), position='dodge') + \
        ggtitle("x: discrete\nposition: 'dodge'"),
    ggplot(data) + geom_errorbar(aes(x='xcat', ymin='ymin', ymax='ymax', color='g'), position=position_dodge(width=.95)) + \
        ggtitle("x: discrete\nposition: position_dodge(width=.95)"),
    ggplot(data) + geom_errorbar(aes(x='xval', ymin='ymin', ymax='ymax', color='g')) + \
        ggtitle("x: continuous\nposition: default"),
    ggplot(data) + geom_errorbar(aes(x='xval', ymin='ymin', ymax='ymax', color='g'), position='dodge') + \
        ggtitle("x: continuous\nposition: 'dodge'"),
    ggplot(data) + geom_errorbar(aes(x='xval', ymin='ymin', ymax='ymax', color='g'), position=position_dodge(width=.95)) + \
        ggtitle("x: continuous\nposition: position_dodge(width=.95)"),
], ncol=3)
```
