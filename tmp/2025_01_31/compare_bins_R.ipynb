```r
library(ggplot2)
library(cowplot)
```

```r
options(repr.plot.width = 12, repr.plot.height = 24)
```

```r
get_plot <- function (geom_kind, bins) {
    p <- ggplot(mpg, aes(cty, hwy))
    if (geom_kind == "bin2d") {
        p <- p + geom_bin2d(bins = bins, color = "black", linewidth = .5, alpha = .5)
    } else if (geom_kind == "hex") {
        p <- p + geom_hex(bins = bins, color = "black", linewidth = .5, alpha = .5)
    }
    p +
        geom_point(color = "red", size = 1.5) +
        scale_fill_continuous(low = "#0c2c84", high = "#ffffcc") +
        coord_cartesian(xlim = c(0, 40), ylim = c(5, 60)) +
        ggtitle(sprintf("%s\nbins=c(%s)", geom_kind, paste(bins, collapse=", ")))
}
```

```r
plot_grid(
    get_plot("bin2d", c(1, 1)), get_plot("hex", c(1, 1)),
    get_plot("bin2d", c(1, 2)), get_plot("hex", c(1, 2)),
    get_plot("bin2d", c(2, 1)), get_plot("hex", c(2, 1)),
    get_plot("bin2d", c(2, 2)), get_plot("hex", c(2, 2)),
    get_plot("bin2d", c(1, 5)), get_plot("hex", c(1, 5)),
    get_plot("bin2d", c(5, 1)), get_plot("hex", c(5, 1)),
    get_plot("bin2d", c(15, 10)), get_plot("hex", c(15, 10)),
    ncol = 2
)
```
