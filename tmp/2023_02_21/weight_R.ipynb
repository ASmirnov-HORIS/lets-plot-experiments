```r
library(ggplot2)
```

```r
df <- data.frame(
    x = c(0, 0, 0, 0, 0, 0, 1, 1, 1),
    g = c("A", "A", "A", "B", "A", "B", "A", "B", "A"),
    w = c(1, 1, 1, 1, -1, -1, 1, 1, -1)
)
```

```r
p <- ggplot(df, aes(x = x, fill = g))
```

```r
# Pure bar
p + geom_bar()
```

```r
# With 'weight'
p + geom_bar(aes(weight = w))
```

```r
# With 'weight' expected
p + geom_bar(aes(group = w, weight = w))
```
