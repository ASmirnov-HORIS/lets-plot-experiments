```r
library(ggplot2)
library(repr)
library(cowplot)
```

```r
options(repr.plot.width=16, repr.plot.height=6)
```

```r
p1 <- ggplot() +
    geom_point(aes(x = 0, y = 0), shape = 1) +
    geom_segment(aes(x = -1, xend = -1, y = 1, yend = -1)) +
    geom_pointrange(aes(x = 1, y = 1, ymin = -1, ymax = 1)) +
    ggtitle("size, stroke, linewidth - default")
p2 <- ggplot() +
    geom_point(aes(x = 0, y = 0), shape = 1, size = 5, stroke = 5) +
    geom_segment(aes(x = -1, xend = -1, y = 1, yend = -1), linewidth = 5) +
    geom_pointrange(aes(x = 1, y = 1, ymin = -1, ymax = 1), shape = 1, size = 5, stroke = 5, linewidth = 5) +
    ggtitle("size, stroke, linewidth = 5")

plot_grid(p1, p2)
```

```r
p1 = ggplot() +
    geom_dotplot(aes(x = 0), binwidth = 1, color = "black", fill = "white") +
    ggtitle("stroke - default")
p2 = ggplot() +
    geom_dotplot(aes(x = 0), binwidth = 1, color = "black", fill = "white", stroke = 5) +
    ggtitle("stroke = 5")

plot_grid(p1, p2)
```

```r
get_data <- function() {
    set.seed(42)
    data.frame(y = rnorm(150))
}

options(repr.plot.width=16, repr.plot.height=4)

df <- get_data()
p1 <- ggplot(df) +
    geom_boxplot(aes(y = y), outlier.shape = 1) +
    ggtitle("default")
p2 <- ggplot(df) +
    geom_boxplot(aes(y = y), outlier.shape = 1, size = 5) +
    ggtitle("size = 5")
p3 <- ggplot(df) +
    geom_boxplot(aes(y = y), outlier.shape = 1, outlier.size = 5, outlier.stroke = 5) +
    ggtitle("outlier.size, outlier.stroke = 5")

plot_grid(p1, p2, p3, ncol = 3)
```
