```python
import matplotlib.pyplot as plt
import numpy as np

X, Y = np.meshgrid(np.arange(3), np.arange(0, 1))
U = X**2
V = Y**2

def get_plot(pivot):
    fig2, ax2 = plt.subplots()
    ax2.set_title("pivot='{0}'".format(pivot))
    Q = ax2.quiver(X, Y, U, V, pivot=pivot)
    ax2.scatter(X, Y, color='r', s=5)
```

```python
get_plot('tail')
```

```python
get_plot('mid')
```

```python
get_plot('middle')
```

```python
get_plot('tip')
```
