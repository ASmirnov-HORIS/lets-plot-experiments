```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
data = {
    'x': [90, 110],
    'y': [0, 0],
    'g': ['a', 'b'],
}
```

```python
p = ggplot(data, aes('x', 'y', color='g')) + geom_livemap(zoom=2) + geom_point() + scale_color_discrete(guide='none') + ggsize(400, 400)
p
```

```python
gggrid([
    p,
    p + facet_grid(x='g'),
    p + facet_grid(y='g'),
], ncol=2)
```
