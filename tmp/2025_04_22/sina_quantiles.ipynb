```python
import numpy as np

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
data = {
    'x': ['a'] * 5,
    'y': [0, 1, 1, 2, 2],
}
seed = 42
ggplot(data, aes("x", "y")) + \
    geom_violin(aes(fill="..quantile.."), quantile_lines=True) + \
    geom_sina(aes(fill="..quantile.."), seed=seed, jitter_y=False, shape=21, color="black") + \
    scale_fill_gradient2(low="red", high="green", mid="yellow", midpoint=.75)
```

```python
data = {
    'x': ['a'] * 4,
    'y': [0, 1, 1, 1],
}
seed = 42
ggplot(data, aes("x", "y")) + \
    geom_violin(aes(fill="..quantile.."), quantile_lines=True, quantiles=[.5]) + \
    geom_sina(aes(fill="..quantile.."), shape=21, color="black", seed=seed, jitter_y=False, quantiles=[.5]) + \
    scale_fill_gradient2(low="red", high="green", mid="yellow", midpoint=.75)
```

```python
np.random.seed(42)

data = {
    "val": np.random.randint(5, size=18),
}

quantiles = [.3, .5, .7]

ggplot(data, aes(y="val")) + \
    geom_violin(aes(fill='..quantile..'), trim=False, quantiles=quantiles, quantile_lines=True) + \
    geom_sina(aes(fill='..quantile..'), color="black", shape=21, jitter_y=False, seed=42, quantiles=quantiles, tooltips=layer_tooltips(["..y..", "..violinwidth.."])) + \
    scale_fill_gradient2(low="red", high="green", mid="yellow", midpoint=.75)
```
