```python
from lets_plot import *
from lets_plot.bistro.waterfall import *
```

```python
LetsPlot.setup_html()
```

```python
data = {
    'accounts': ["initial", "revenue", "costs", "result"],
    'values': [20, 20, -10, None],
    'measure': ['absolute', 'relative', 'relative', 'total'],
}
custom_scale = scale_manual(['color', 'fill'], {"Absolute": "pacificblue", "Increase": "green", "Decrease": "red", "Total": "blue"})
```

```python
gggrid([
    waterfall_plot(data, 'accounts', 'values', measure='measure', show_legend=True) + custom_scale,
    waterfall_plot(data, 'accounts', 'values', measure='measure', show_legend=True, size=.75, color="black") + custom_scale,
    waterfall_plot(data, 'accounts', 'values', measure='measure', show_legend=True, size=.75, color="black", label={'color': 'flow_type'}, alpha=.25) + custom_scale,
    waterfall_plot(data, 'accounts', 'values', measure='measure', show_legend=True, size=.75, color='flow_type', fill="flow_type", alpha=.5) + custom_scale,
    waterfall_plot(data, 'accounts', 'values', measure='measure', show_legend=True, size=.75, color='flow_type', fill="transparent", label={'color': "black"}, linetype='dashed') + custom_scale,
    waterfall_plot(data, 'accounts', 'values', measure='measure', show_legend=True, size=.75, color="black", fill="transparent", label={'color': 'flow_type'}) + custom_scale,
    waterfall_plot(data, 'accounts', 'values', measure='measure', show_legend=True, size=.75, color='flow_type', fill="rgba(128,0,128,0.5)") + custom_scale,
    waterfall_plot(data, 'accounts', 'values', measure='measure', show_legend=True, size=.75, color='flow_type', fill="rgba(128,0,128,0.5)", label={'color': 'flow_type'}) + custom_scale,
    waterfall_plot(data, 'accounts', 'values', measure='measure', show_legend=True, size=.75, color="black", fill="rgba(128,0,128,0.5)") + custom_scale,
], ncol=3)
```
