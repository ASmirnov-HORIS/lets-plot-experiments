```python
import pandas as pd

from lets_plot import *
```

```python
LetsPlot.setup_html(no_js=True)
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/refs/heads/master/data/mpg.csv")
print(df.shape)
df.head()
```

```python
gggrid([
    ggplot(df, aes("drv", "hwy")) + \
        geom_violin() + \
        geom_sina(aes(color="hwy"), seed=42) + \
        ggtitle('color="hwy"\nContinuous, same as y'),
    ggplot(df, aes("drv", "hwy")) + \
        geom_violin() + \
        geom_sina(aes(color="drv"), seed=42) + \
        ggtitle('color="drv"\nDiscrete, same as x'),
    ggplot(df, aes("drv", "hwy")) + \
        geom_violin() + \
        geom_sina(aes(color="cyl"), seed=42) + \
        ggtitle('color="cyl"\nContinuous, differs from x/y'),
    ggplot(df, aes("drv", "hwy")) + \
        geom_violin() + \
        geom_sina(aes(color="fl"), seed=42) + \
        ggtitle('color="fl"\nDiscrete, differs from x/y'),
    ggplot(df, aes("drv", "hwy")) + \
        geom_violin() + \
        geom_sina(aes(color="cyl"), seed=42) + \
        scale_color_brewer(type='qual', breaks=sorted(df["cyl"].unique())) + \
        ggtitle('color="cyl"\nContinuous, differs from x/y, custom discrete palette'),
    ggplot(df, aes("drv", "hwy")) + \
        geom_violin() + \
        geom_sina(aes(color="fl", group="drv"), seed=42) + \
        ggtitle('color="fl"\nDiscrete, differs from x/y, grouping by "drv"'),
], ncol=2)
```
