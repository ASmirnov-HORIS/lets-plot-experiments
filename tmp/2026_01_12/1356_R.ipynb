```r
library(ggplot2)
```

```r
df1 <- data.frame(
    x = c(0, 1, 2, 0, 1, 2, 0, 1, 2),
    y = c(0, 0, 0, 1, 1, 1, 2, 2, 2),
    z = c(0, 1, 2, 3, 4, 5, 6, 7, 8)
)

ggplot(df1) + geom_contour(aes(x, y, z = z))
```
