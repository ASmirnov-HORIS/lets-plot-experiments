```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
def get_data(*, a=1, n):
    return [a * 10**d for d in range(-n, n + 1)]
```

```python
def get_plot(*, a=1, n):
    v = get_data(a=a, n=n)
    return ggplot() + \
        geom_text(aes(y=v, label=v), label_format='~e') + \
        scale_y_log10(limits=[10**(-n - 1), 10**(n + 1)])
```

```python
gggrid([
    get_plot(n=5),
    get_plot(n=5) + theme(exponent_format='pow'),
    get_plot(a=2, n=5),
    get_plot(a=2, n=5) + theme(exponent_format='pow'),
], ncol=2)
```
