```r
library(ggplot2)
library(ggpointdensity)
```

```r
df <- data.frame(
    x = c(0, 0, 1),
    y = c(0, 1, 0)
)
```

```r
ggplot(df, aes(x, y)) + stat_pointdensity(aes(label=after_stat(density)), geom='text', adjust=0)
```

```r
ggplot(df, aes(x, y)) + stat_pointdensity(aes(label=after_stat(density)), geom='text', adjust=69.999)
```

```r
ggplot(df, aes(x, y)) + stat_pointdensity(aes(label=after_stat(density)), geom='text', adjust=70)
```
