```r
library(ggplot2)
```

```r
df <- data.frame(
    x = c(0, 1, 2, 3, 4),
    y = c(0, 0, 0, 0, 0),
    s = c(2, 4, 32, 8, 16),
    cs = c('a', 'b', 'c', 'd', 'e')
)
df
```

```r
ggplot(df, aes(x, y)) + geom_point(aes(size = s), shape = 1)
```

```r
ggplot(df, aes(x, y)) + geom_point(aes(size = s, stroke = s), shape = 1)
```

```r
ggplot(df, aes(x, y)) + geom_point(aes(size = s), shape = 1) + scale_size_identity()
```

```r
ggplot(df, aes(x, y)) + geom_point(aes(size = s, stroke = s), shape = 1) + scale_size_identity()
```

```r
ggplot(df, aes(x, y)) + geom_point(aes(size = cs), shape = 1)
```

```r
ggplot(df, aes(x, y)) + geom_point(aes(size = cs, stroke = cs), shape = 1)
```
