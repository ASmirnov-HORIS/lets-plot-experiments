```r
p <- ggplot(mpg, aes(drv, cty, color = drv, fill = drv)) +
    geom_boxplot(alpha = .1) +
    geom_point(shape = 21, size=3, fill = "yellow")
p
```

```r
ggMarginal(p, margins = 'y', type = "violin", groupFill = TRUE)
```

```r
ggMarginal(p, margins = 'y', type = "histogram", groupFill = TRUE)
```
