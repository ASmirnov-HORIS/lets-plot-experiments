## Extra `y` aesthetic breaks the `geom_density()`

```r
library(ggplot2)
```

### `geom_density()`

```r
ggplot(mpg, aes(x = cty, fill = drv)) + geom_density()
```

```r
ggplot(mpg, aes(cty, drv, fill = drv)) + geom_density()
```

```r
ggplot(mpg, aes(cty, drv, fill = drv)) + geom_density(aes(group = drv))
```

### `geom_histogram()`

```r
ggplot(mpg, aes(x = cty, fill = drv)) + geom_histogram()
```

```r
ggplot(mpg, aes(cty, drv, fill = drv)) + geom_histogram()
```

```r
ggplot(mpg, aes(cty, drv, fill = drv)) + geom_histogram(aes(group = drv))
```

### `geom_linerange()`

```r
df <- data.frame(
    x = c(-2, 2),
    y = c(-1, 1),
    ymin = c(-1.5, -.5),
    ymax = c(.5, 1.5)
)

ggplot(df, aes(x, y, ymin = ymin, ymax = ymax)) +
    geom_linerange() # Can work with aes y, but don't use it
```
