## Extra `y` aesthetic breaks the `geom_density()`

```python
import pandas as pd

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/refs/heads/master/data/mpg.csv")
print(df.shape)
df.head()
```

### `geom_density()`

```python
ggplot(df, aes(x="cty", fill="drv")) + geom_density()
```

```python
ggplot(df, aes("cty", "drv", fill="drv")) + geom_density()
```

```python
ggplot(df, aes("cty", "drv", fill="drv")) + geom_density(aes(group="drv"))
```

### `geom_histogram()`

```python
ggplot(df, aes(x="cty", fill="drv")) + geom_histogram()
```

```python
ggplot(df, aes("cty", "drv", fill="drv")) + geom_histogram()
```

```python
ggplot(df, aes("cty", "drv", fill="drv")) + geom_histogram(aes(group="drv"))
```

### `geom_linerange()`

```python
data = {
    'x': [-2, 2],
    'y': [-1, 1],
    'ymin': [-1.5, -.5],
    'ymax': [.5, 1.5],
}

ggplot(data, aes('x', 'y', ymin='ymin', ymax='ymax')) + \
    geom_linerange() # Can work with aes y, but don't use it
```
