```r
library(ggplot2)
```

```r
df <- data.frame(
    x = c(1, 2, 3,
          1, 1),
    y = c(1, 1, 1,
          3, 3),
    shape = c(2, 6, 0,
              2, 6)
)
```

```r
ggplot(df, aes(x, y)) +
    geom_point(aes(shape = shape, color = as.factor(shape)), size = 25, stroke = 5) +
    geom_point(aes(color = as.factor(shape)), size = 2) +
    scale_shape_identity() +
    xlim(0, 4) + ylim(0, 4)
```
