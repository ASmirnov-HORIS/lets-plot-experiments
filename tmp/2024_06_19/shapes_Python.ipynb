```python
from lets_plot import *
from lets_plot.mapping import as_discrete
```

```python
LetsPlot.setup_html()
```

```python
data = {
    'x': [1, 2, 3, \
          1, 1, 2, 3],
    'y': [1, 1, 1, \
          3, 3, 3, 3],
    'shape': [2, 6, 0, \
              2, 6, 2, 2],
    'angle': [0, 0, 0, \
              0, 0, 30, 90],
}
```

```python
ggplot(data, aes('x', 'y')) + \
    geom_point(aes(shape='shape', angle='angle', color=as_discrete('shape')), size=30, stroke=5, alpha=.5) + \
    geom_point(aes(color=as_discrete('shape')), size=2, alpha=.5) + \
    scale_shape_identity() + \
    xlim(0, 4) + ylim(0, 4)
```
