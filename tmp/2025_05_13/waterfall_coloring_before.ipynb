```python
data = {
    'accounts': ["initial", "revenue", "costs", "Q1", "revenue", "costs", "Q2"],
    'values': [200, 200, -100, None, 250, -100, None],
    'measure': ['absolute', 'relative', 'relative', 'total', 'relative', 'relative', 'total'],
}
```

```python
gggrid([
    waterfall_plot(data, 'accounts', 'values', measure='measure', show_legend=True),
    waterfall_plot(data, 'accounts', 'values', measure='measure', show_legend=True, size=.75, color="black"),
    waterfall_plot(data, 'accounts', 'values', measure='measure', show_legend=True, size=.75, color='flow_type', fill="gray"),
    waterfall_plot(data, 'accounts', 'values', measure='measure', show_legend=True, size=.75, color="black", fill="white", label={'color': 'flow_type'}),
    waterfall_plot(data, 'accounts', 'values', measure='measure', show_legend=True, size=.75, color="black", label={'color': 'flow_type'}, alpha=.25),
    waterfall_plot(data, 'accounts', 'values', measure='measure', show_legend=True, size=.75, color='flow_type', fill="transparent", label={'color': "black"}, linetype='dotted'),
], ncol=2)
```
