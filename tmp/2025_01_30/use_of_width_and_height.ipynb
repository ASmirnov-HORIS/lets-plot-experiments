```r
library(ggplot2)
```

```r
get_plot <- function (width, height) {
    ggplot() +
        geom_hex(aes(x = 0, y = 0, width = width, height = height), stat = "identity") +
        coord_fixed(ratio = 1, xlim = c(-2, 2), ylim = c(-2, 2)) +
        ggtitle(sprintf("width = %s, height = %s", width, height))
}
```

```r
get_plot(1, 1)
```

```r
get_plot(1, 2)
```

```r
get_plot(3, 2)
```
