```r
library(ggplot2)
```

```r
h = 2 / sqrt(3)
```

```r
get_plot <- function (cx, cy) {
    df <- data.frame(
        x = c(-cx / 2, cx / 2, 0),
        y = c(0, 0, 2 * h * cy)
    )
    ggplot(df, aes(x, y)) + geom_hex(binwidth = c(cx, cy), color = "white", linewidth = .25) + geom_point(color = "red")
}
```

```r
get_plot(1, 1)
```

```r
get_plot(1e1, 1e1)
```

```r
get_plot(1e-1, 1e-1)
```

```r
get_plot(1e8, 1e8)
```

```r
get_plot(1e-8, 1e-8)
```
