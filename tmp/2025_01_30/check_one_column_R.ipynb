```r
library(ggplot2)
```

```r
get_plot <- function (bins) {
    ggplot(mpg, aes(cty, hwy)) +
        geom_hex(bins = bins, color = "white", linewidth = .25) +
        geom_point(color = "red") +
        ggtitle(sprintf("bins = c(%s)", paste(bins, collapse=", ")))
}
```

```r
get_plot(c(1, 1))
```

```r
get_plot(c(1, 2))
```

```r
get_plot(c(2, 1))
```

```r
get_plot(c(2, 2))
```

```r
get_plot(c(1, 5))
```

```r
get_plot(c(10, 10))
```
