# Multiple Color Scales

```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
def generate_data(p1, p2, p3, p4, x_max=400, y_max=300):
    import math

    def calculate_z(x, y):
        z = math.sin(x * p1 * math.pi / x_max)
        z += math.sin(y * p2 * math.pi / y_max)
        z += x * p3 / x_max
        z += y * p4 / y_max
        return z

    x = []
    y = []
    z = []
    for row in range(y_max - 1):
        for col in range(x_max - 1):
            x.append(col)
            y.append(row)
            z.append(calculate_z(col, row))

    return dict(x=x, y=y, z=z)
```

```python
height_data = generate_data(3, 3, 3, 5)
temperature_data = generate_data(1, 2, 5, 4)
```

```python
ggplot() + \
    geom_contour(aes('x', 'y', z='z', paint_a='..level..'), \
                 data=height_data, color_by='paint_a', \
                 bins=15, size=1) + \
    geom_contour(aes('x', 'y', z='z', paint_b='..level..'), \
                 data=temperature_data, color_by='paint_b', \
                 bins=8, size=1) + \
    scale_gradient('paint_a', name="height", low="#ffffd4", high="#993404") + \
    scale_gradient('paint_b', name="temperature", \
                   low="#0571b0", high="#ca0020") + \
    ggsize(800, 600)
```
