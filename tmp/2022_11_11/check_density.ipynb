```r
library(ggplot2)
```

```r
df <- data.frame(
    x = c(0.0, 1.0, 1.0, 2.0, 0.0, 0.0, 1.0, 2.0),
    g = c("A", "A", "A", "A", "B", "B", "B", "B")
)
df
```

```r
ggplot(df, aes(x = x)) +
    geom_area(aes(y = ..density.., fill = g), stat = "density", position = "identity", color = "black", alpha = 0.5) +
    ggtitle("y = ..density.., position = identity")
```

```r
ggplot(df, aes(x = x)) +
    geom_area(aes(y = ..density.., fill = g), stat = "density", position = "stack", color = "black", alpha = 0.5) +
    ggtitle("y = ..density.., position = stack")
```

```r
ggplot(df, aes(x = x)) +
    geom_area(aes(y = ..count.., fill = g), stat = "density", position = "identity", color = "black", alpha = 0.5) +
    ggtitle("y = ..count.., position = identity")
```

```r
ggplot(df, aes(x = x)) +
    geom_area(aes(y = ..count.., fill = g), stat = "density", position = "stack", color = "black", alpha = 0.5) +
    ggtitle("y = ..count.., position = stack")
```
