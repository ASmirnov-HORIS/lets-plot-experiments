```kotlin
%useLatestDescriptors
%use lets-plot
```

```kotlin
val p1 = letsPlot() + geomLine() { x = listOf(-1, -1, 1, 1); y = listOf(-1, 1, 1, -1) }
ggsave(p1, "plot1.png")
```

```kotlin
val p2 = letsPlot() + geomLine(size = 5) { x = listOf(-1, -1, 1, 1); y = listOf(-1, 1, 1, -1) }
ggsave(p2, "plot2.png")
```

```kotlin
ggsave(p2 + geomPoint(x = -1, y = -1, shape = 1, size = 10, color = "red", stroke = 2), "example.png")
```
