```r
library(ggplot2)
library(cowplot)
```

```r
get_plot <- function (xs = NULL, binwidth = NULL, center = NULL, boundary = NULL) {
    df <- data.frame(x = if (is.null(xs)) c(5, 5, 5) else xs)
    ggplot(df, aes(x = x)) + geom_histogram(binwidth = binwidth, center = center, boundary = boundary)
}

get_plot1 <- function (binwidth = NULL, center = NULL, boundary = NULL) {
    get_plot(c(4, 5, 5), binwidth = binwidth, center = center, boundary = boundary)
}
```

```r
options(repr.plot.width = 16, repr.plot.height = 16)
```

```r
plot_grid(
    get_plot(),
    get_plot(binwidth = .01),
    get_plot(binwidth = 100),
    get_plot(center = 5),
    get_plot(center = 4.75),
    get_plot(center = 4.5),
    get_plot(center = .5),
    get_plot(center = .25),
    get_plot(center = 10.25),
    get_plot(boundary = 5),
    get_plot(boundary = 4.75),
    get_plot(boundary = 4.5),
    get_plot(boundary = .5),
    get_plot(boundary = .25),
    get_plot(boundary = 10.25),
    ncol = 3
)
```

```r
plot_grid(
    get_plot1(),
    get_plot1(binwidth = .01),
    get_plot1(binwidth = 100),
    get_plot1(center = 5),
    get_plot1(center = 4.75),
    get_plot1(center = 4.5),
    get_plot1(center = .5),
    get_plot1(center = .25),
    get_plot1(center = 10.25),
    get_plot1(boundary = 5),
    get_plot1(boundary = 4.75),
    get_plot1(boundary = 4.5),
    get_plot1(boundary = .5),
    get_plot1(boundary = .25),
    get_plot1(boundary = 10.25),
    ncol = 3
)
```
