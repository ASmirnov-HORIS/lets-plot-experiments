```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
def get_plot(xs=None, binwidth=None, center=None, boundary=None):
    data = {
        'x': [5, 5, 5] if xs is None else xs
    }
    return ggplot(data, aes(x='x')) + \
        geom_histogram(binwidth=binwidth, center=center, boundary=boundary) + \
        ggtitle(f"binwidth={binwidth}\ncenter={center}, boundary={boundary}")

def get_plot1(binwidth=None, center=None, boundary=None):
    return get_plot([4, 5, 5], binwidth=binwidth, center=center, boundary=boundary)
```

```python
gggrid([
    get_plot(),
    get_plot(binwidth=.1),
    get_plot(binwidth=10),
    get_plot(center=5),
    get_plot(center=4.75),
    get_plot(center=4.5),
    get_plot(center=.5),
    get_plot(center=.25),
    get_plot(center=10.25),
    get_plot(boundary=5),
    get_plot(boundary=4.75),
    get_plot(boundary=4.5),
    get_plot(boundary=.5),
    get_plot(boundary=.25),
    get_plot(boundary=10.25),
], ncol=3)
```

```python
gggrid([
    get_plot1(),
    get_plot1(binwidth=.1),
    get_plot1(binwidth=10),
    get_plot1(center=5),
    get_plot1(center=4.75),
    get_plot1(center=4.5),
    get_plot1(center=.5),
    get_plot1(center=.25),
    get_plot1(center=10.25),
    get_plot1(boundary=5),
    get_plot1(boundary=4.75),
    get_plot1(boundary=4.5),
    get_plot1(boundary=.5),
    get_plot1(boundary=.25),
    get_plot1(boundary=10.25),
], ncol=3)
```
