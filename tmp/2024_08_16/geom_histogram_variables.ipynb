```r
library(ggplot2)
```

```r
df <- data.frame(x = c(0, 2, 2, 4))
```

```r
g <- ggplot(df, aes(x = x))
```

```r
# count: 1, 2, 1
g + geom_histogram(binwidth = 2)
```

```r
# density: 0.125, 0.25, 0.125
# area: 2 * (0.125 + 0.25 + 0.125) = 1
g + geom_histogram(aes(y = after_stat(density)), binwidth = 2)
```

```r
# count: 1, 2, 1
# max count: 2
# ncount (= count / max count): 0.5, 1, 0.5
# max ncount: 1
g + geom_histogram(aes(y = after_stat(ncount)), binwidth = 2)
```

```r
# density: 0.125, 0.25, 0.125
# max density: 0.25
# ndensity (= density / max density): 0.5, 1, 0.5
# max ndensity: 1
g + geom_histogram(aes(y = after_stat(ndensity)), binwidth = 2)
```

```r
# case with different values for ncount and ndensity
ggplot(data.frame(x = rnorm(1000)), aes(x = x)) +
    geom_histogram(aes(y = after_stat(ncount)), breaks = c(-3.5, -3, -2, 0, 2, 3, 3.5),
                   color = "black", fill = "red", alpha = 0.25) +
    geom_histogram(aes(y = after_stat(ndensity)), breaks = c(-3.5, -3, -2, 0, 2, 3, 3.5),
                   color = "black", fill = "blue", alpha = 0.25)
```
