```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
def get_data1(n):
    data = {
        'x': [i for i in range(n) for j in range(n)],
        'y': [j for i in range(n) for j in range(n)],
    }
    data['z'] = [x + y for x, y in zip(data['x'], data['y'])]
    return data

def get_plot1(n):
    return ggplot(get_data1(n), aes('x', 'y', fill='z')) + \
        geom_raster(show_legend=False) + \
        theme_void()
```

```python
get_plot1(100)
```

```python
def get_data2():
    return {
        'x': [0, 0],
        'y': [0, 1],
        'l': ["T1", "T2"],
    }

def get_plot2():
    return ggplot(get_data2(), aes('x', 'y')) + \
        geom_label(aes(label='l', color='l'))
```

```python
get_plot2()
```
