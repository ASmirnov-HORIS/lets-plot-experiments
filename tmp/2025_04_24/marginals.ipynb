```python
import numpy as np

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
def get_data(n=100, m=2, seed=42):
    np.random.seed(seed)
    return {
        'x': [int(v * m) / m for v in np.random.normal(size=n)],
        'y': [int(v * m) / m for v in np.random.normal(size=n)],
    }

data = get_data()
```

```python
g = ggplot(data, aes('x', 'y')) + \
    geom_point(alpha=.25, size=5)
g
```

```python
gggrid([
    g + \
        ggmarginal("tr", layer=stat_summary()) + \
        ggtitle("stat_summary()"),
    g + \
        ggmarginal("t", layer=stat_summary(orientation='y')) + \
        ggmarginal("r", layer=stat_summary(orientation='x')) + \
        ggtitle("stat_summary() with explicit orientation"),
    g + \
        ggmarginal("tr", layer=geom_pointrange(stat='boxplot')) + \
        ggtitle("geom_pointrange(stat='boxplot')"),
])
```

```python
gggrid([
    g + \
        ggmarginal("tr", layer=geom_ydotplot()),
    g + \
        ggmarginal("t", layer=geom_ydotplot(orientation='y')) + \
        ggmarginal("r", layer=geom_ydotplot(orientation='x')),
])
```
