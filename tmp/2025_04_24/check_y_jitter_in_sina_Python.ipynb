```python
def get_plot(values):
    return ggplot({'v': values}, aes(y='v')) + \
        geom_violin() + \
        geom_point(size=2, shape=21, stroke=.25, fill="yellow") + \
        geom_sina(seed=42, size=1, shape=21, stroke=.25, fill="orange")
```

```python
get_plot([1, 1, 2, 2, 2])
```

```python
get_plot([10, 10, 20, 20, 20])
```

```python
import numpy as np
```
