```r
library(ggplot2)
library(ggforce)
```

```r
get_plot <- function (values) {
    ggplot(data.frame(v = values), aes(x = 0, y = v)) +
        geom_violin() +
        geom_point(size = 4, shape = 21, fill = "yellow") +
        geom_sina(seed = 42, size = 2, shape = 21, fill = "orange")
}
```

```r
get_plot(c(1, 1, 2, 2, 2))
```

```r
get_plot(c(10, 10, 20, 20, 20))
```

```r
get_plot((rpois(10000, 1) + 1) * 2)
```

```r
get_plot((rpois(10000, 1) + 1) * 10)
```

```r
get_plot((rpois(10000, 1) + 1) * 1000)
```
