```python
import numpy as np
import pandas as pd

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
def get_data(size=100, seed=42):
    return pd.DataFrame({
        'cat': np.random.choice(['a', 'b', 'c'], size=size),
        'val': np.random.normal(size=size),
        'group': np.random.choice(['g1', 'g2'], size=size),
    })

df = get_data()
print(df.shape)
df.head()
```

```python
ggplot(df, aes(x='val', y='cat', color='group')) + \
    geom_errorbar(stat='summary', position=position_dodge(width=.5)) + \
    geom_point(position=position_dodgev(height=.5))
```
