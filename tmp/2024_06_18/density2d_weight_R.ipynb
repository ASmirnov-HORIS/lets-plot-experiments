```r
library(ggplot2)
library(cowplot)
```

```r
options(repr.plot.width=16, repr.plot.height=6)
```

```r
df1 <- data.frame(
    x = c(0, 1, 1),
    y = c(0, 1, 1)
)
df2 <- data.frame(
    x = c(0, 1),
    y = c(0, 1),
    w = c(1, 2)
)
```

```r
plot_grid(
    ggplot() +
        geom_density2d(aes(x, y), data = df1, color = "blue") +
        geom_density2d(aes(x, y, weight = w), data = df2, color = "red") +
        ggtitle("bw is default, adjust is default\nblue - without weight, red - with weight"),
    ggplot() +
        geom_density2d(aes(x, y), data = df1, h = c(1, 1), adjust = 1, color = "blue") +
        geom_density2d(aes(x, y, weight = w), data = df2, h = c(1, 1), adjust = 1, color = "red") +
        ggtitle("bw is default, adjust is default\nblue - without weight, red - with weight")
)
```
