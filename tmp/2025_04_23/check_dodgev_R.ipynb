```r
library(ggplot2)
```

```r
df <- data.frame(
    valMin = c(-2, -3),
    valMax = c(3, 2),
    group = c('A', 'B')
)
```

```r
ggplot(df) + geom_errorbar(aes(x = 0, ymin = valMin, ymax = valMax, color = group), linewidth = 2, position = position_dodge())
```

```r
ggplot(df) + geom_errorbar(aes(y = 0, xmin = valMin, xmax = valMax, color = group), linewidth = 2, position = position_dodge())
```

```r
ggplot(df) + geom_errorbar(aes(y = 0, xmin = valMin, xmax = valMax, color = group), linewidth = 2, position = "dodge")
```

```r
ggplot(df) + geom_errorbar(aes(y = 0, xmin = valMin, xmax = valMax, color = group), linewidth = 2, position = position_dodgev())
```

```r
ggplot(df) + geom_errorbar(aes(y = 0, xmin = valMin, xmax = valMax, color = group), linewidth = 2, position = "dodgev")
```
