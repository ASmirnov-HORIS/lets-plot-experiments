```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
data = {
    'x': [10, 10],
    'y': [9999, 9999],
    'g': ['A', 'B'],
}
```

```python
g = ggplot(data, aes('x', 'y', color='g'))
gggrid([
    g + geom_point(size=10, alpha=.5) + ggtitle("Position identity"),
    g + geom_point(size=10, alpha=.5, position='dodge') + ggtitle("Position dodge"),
    g + geom_point(size=10, alpha=.5, position='dodgev') + ggtitle("Position dodgev"),
])
```
