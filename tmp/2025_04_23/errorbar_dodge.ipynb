```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
data1 = {
    'xcat': ['x', 'x'],
    'xval': [0, 0],
    'ymin': [-2, -1],
    'ymax': [1, 2],
    'g': ['a', 'b'],
}
```

```python
p1 = ggplot(data1) + geom_errorbar(aes(x='xcat', ymin='ymin', ymax='ymax', color='g')) + \
    ggtitle("x: discrete\nposition: default")
p1
```

```python
p2 = ggplot(data1) + geom_errorbar(aes(x='xcat', ymin='ymin', ymax='ymax', color='g'), position='dodge') + \
    ggtitle("x: discrete\nposition: 'dodge'")
p2
```

```python
p3 = ggplot(data1) + geom_errorbar(aes(x='xcat', ymin='ymin', ymax='ymax', color='g'), position=position_dodge(width=.45)) + \
    ggtitle("x: discrete\nposition: position_dodge(width=.45)")
p3
```

```python
p4 = ggplot(data1) + geom_errorbar(aes(x='xval', ymin='ymin', ymax='ymax', color='g')) + \
    ggtitle("x: continuous\nposition: default")
p4
```

```python
p5 = ggplot(data1) + geom_errorbar(aes(x='xval', ymin='ymin', ymax='ymax', color='g'), position='dodge') + \
    ggtitle("x: continuous\nposition: 'dodge'")
p5
```

```python
p6 = ggplot(data1) + geom_errorbar(aes(x='xval', ymin='ymin', ymax='ymax', color='g'), position=position_dodge(width=.45)) + \
    ggtitle("x: continuous\nposition: position_dodge(width=.45)")
p6
```

```python
gggrid([p1, p2, p3, p4, p5, p6], ncol=3) + ggsize(800, 500)
```
