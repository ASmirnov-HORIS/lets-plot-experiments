```r
library(ggplot2)
```

```r
df <- data.frame(
    decade = c(1990, 2000, 2000)
)
```

## Default

```r
ggplot(df) +
    geom_bar(aes(x = decade))
```

## With `scale_x_discrete()`

```r
ggplot(df) +
    geom_bar(aes(x = decade)) +
    scale_x_discrete(name = "Decade")
```

## With `scale_x_discrete()` and `as.factor()`

```r
ggplot(df) +
    geom_bar(aes(x = as.factor(decade))) +
    scale_x_discrete(name = "Decade")
```
