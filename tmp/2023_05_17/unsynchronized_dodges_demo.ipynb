```python
from lets_plot import *
LetsPlot.setup_html(offline=True, no_js=True)
```

```python
boxes_data = dict(
    x = [0, 1, 1],
    ymin = [-3, -1, -1],
    lower = [-2, 0, 0],
    middle = [-1, 1, 1],
    upper = [0, 2, 2],
    ymax = [1, 3, 3],
    g = ['a', 'b', 'a']
)
points_data = dict(
    x = [0, 1, 1],
    y = [2, -2, -3],
    g = ['a', 'b', 'a']
)
ggplot() + \
    geom_boxplot(aes('x', ymin='ymin', lower='lower', middle='middle', upper='upper', ymax='ymax', color='g'), \
                 data=boxes_data, stat='identity', position='dodge') + \
    geom_point(aes('x', 'y', color='g'), data=points_data, position=position_dodge(width=.95), size=5) + \
    ggtitle("Одинаковые x и группы у ящиков и точек")
```
