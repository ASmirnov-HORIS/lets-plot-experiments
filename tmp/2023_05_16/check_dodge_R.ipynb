```r
library(ggplot2)
```

```r
boxes_df <- data.frame(
    x = c(0, 1),
    ymin = c(-3, -1),
    lower = c(-2, 0),
    middle = c(-1, 1),
    upper = c(0, 2),
    ymax = c(1, 3),
    g = c('a', 'b')
)
points_df <- data.frame(
    x = c(0, 1),
    y = c(2, -2),
    g = c('a', 'b')
)
ggplot() +
    geom_boxplot(aes(x, ymin = ymin, lower = lower, middle = middle, upper = upper, ymax = ymax, color = g),
                 data = boxes_df, stat = 'identity', position = 'dodge') +
    geom_point(aes(x, y, color = g), data = points_df, position = position_dodge(width = 0.4), size = 5) +
    ggtitle("position='dodge'")
```
