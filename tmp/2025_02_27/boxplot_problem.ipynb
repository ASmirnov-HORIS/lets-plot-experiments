```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
data = {
    'cat': ['a', 'b'],
    'min': [-3, -2],
    'lower': [-2, -1],
    'upper': [1, 2],
    'max': [2, 3]
}
```

```python
ggplot(data) + geom_boxplot(aes(x='cat', ymin='min', lower='lower', upper='upper', ymax='max'), stat='identity')
```

```python
ggplot(data) + geom_boxplot(aes(y='cat', xmin='min', lower='lower', upper='upper', xmax='max'), stat='identity')
```
