```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
data = {
    'xmin': [-2, -1],
    'xmax': [1, 2],
    'y': ['a', 'b'],
}
```

```python
def get_plot(geom, **options):
    return ggplot(data) + \
        geom(aes(xmin='xmin', xmax='xmax', y='y'), **options) + \
        ggtitle("Options: {0}".format(options))
```

```python
gggrid([
    get_plot(geom_errorbar),
    get_plot(geom_errorbar, width=.1),
    get_plot(geom_errorbar, height=.1),
])
```

```python
gggrid([
    get_plot(geom_crossbar),
    get_plot(geom_crossbar, width=.1),
    get_plot(geom_crossbar, height=.1),
])
```
