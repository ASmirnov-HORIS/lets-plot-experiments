```python
import pandas as pd

from lets_plot import *
from lets_plot.bistro import *
```

```python
LetsPlot.setup_html()
LetsPlot.set_theme(theme(plot_title=element_text(size=18, face='bold')))
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv")
df.head()
```

```python
# colored
# combination of geoms on one marginal
# repetitions in sides and geoms
# spaces between values
# invalid values
# empty and small datasets
gggrid([
    qq_plot(df, "hwy") + ggtitle("Default", "sample vs. theoretical"),
    qq_plot(df, x="cty", y="hwy") + ggtitle("Default", "x vs. y"),
    qq_plot(df, "hwy", marginal="hist:tr") + ggtitle("Histogram", "sample vs. theoretical"),
    qq_plot(df, x="cty", y="hwy", marginal="hist:tr") + ggtitle("Histogram", "x vs. y"),
    qq_plot(df, "hwy", marginal="box:tr") + ggtitle("Boxplot", "sample vs. theoretical"),
    qq_plot(df, x="cty", y="hwy", marginal="box:tr") + ggtitle("Boxplot", "x vs. y"),
    qq_plot(df, "hwy", marginal='none') + ggtitle("Without marginals", "sample vs. theoretical"),
    qq_plot(df, x="cty", y="hwy", marginal='none') + ggtitle("Without marginals", "x vs. y"),
    qq_plot(df, "hwy", marginal="dens:tr:.3") + ggtitle("Specified size", "sample vs. theoretical"),
    qq_plot(df, x="cty", y="hwy", marginal="dens:tr:.3") + ggtitle("Specified size", "x vs. y"),
    qq_plot(df, "hwy", marginal="density:t,histogram:r,boxplot:bl") + ggtitle("Different sides and full names", "sample vs. theoretical"),
    qq_plot(df, x="cty", y="hwy", marginal="density:t,histogram:r,boxplot:bl") + ggtitle("Different sides and full names", "x vs. y"),
], ncol=2)
```
