```r
library(ggplot2)
```

```r
bar_data <- data.frame(x = c("A", "B", "C"), y = c(1, 3, 2))
```

```r
ggplot(bar_data, aes(x = x, y = y)) +
  geom_bar(stat = "identity", width = 0.5) +
  ggtitle("Bar Chart: Width = 0.5 of X spacing")
```

```r
ggplot(bar_data, aes(x = x, y = y)) +
  geom_bar(stat = "identity", width = 0.1) +
  ggtitle("Bar Chart: Width = 0.5 of Y units")
```

```r
bin2d_data <- data.frame(x = c(0, 0, 2, 2), y = c(0, 2, 0, 0))
```

```r
ggplot(bin2d_data, aes(x = x, y = y)) +
  stat_bin2d(binwidth = c(1, 1), drop = TRUE, linewidth = 1, color = "white", width = 1, height = 1) +
  scale_fill_viridis_c()
```

```r
bin2d_identity_data <- data.frame(x = c(0, 0, 2), y = c(0, 2, 0), c = c(1, 1, 2))
```

```r
ggplot(bin2d_identity_data, aes(x = x, y = y)) +
  geom_bin2d(aes(fill = c), stat = "identity", linewidth = 1, color = "white", width = 1, height = 1) +
  scale_fill_viridis_c()
```
