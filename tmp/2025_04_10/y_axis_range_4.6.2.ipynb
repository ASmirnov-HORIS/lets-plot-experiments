```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
data = dict(
    cat = ["a", "b", "c"],
    l = [20, 5, 30],
    u = [27, 10, 39],
    m = [25, 7, 32],
)

gggrid([
    ggplot(data) + \
        geom_boxplot(aes(x='cat', lower='l', middle='m', upper='u'), stat='identity') + \
        theme_bw() + \
        ggtitle("geom_boxplot()\nCorrect y-axis range"),
    ggplot(data) + \
        geom_crossbar(aes(x='cat', ymin='l', y='m', ymax='u')) + \
        theme_bw() + \
        ggtitle("geom_crossbar()\nIncorrect y-axis range"),
]) + ggsize(600, 600)
```
