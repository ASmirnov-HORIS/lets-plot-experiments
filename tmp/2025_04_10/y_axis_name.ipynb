```python
data = {
    'cat': ['a'] * 3 + ['b'] * 3,
    'val': [99, 100, 102, 98, 100, 101],
}

p = ggplot(data, aes('cat', 'val')) + theme(axis_title_y=element_text(color='red'))

gggrid([
    p + geom_boxplot(),
    p + layer(geom='boxplot'),
])
```
