```r
library(ggplot2)
```

```r
hh = 2 / sqrt(3)
```

```r
df <- data.frame(
    x = c(-1, -1, 1, 1, 1),
    y = c(-1, 1, -1, 1, 1)
)
```

```r
ggplot(df, aes(x, y)) + geom_hex(bins=2)
```

```r
ggplot(df, aes(x, y)) + geom_hex(binwidth=1)
```

```r
p <- ggplot(df, aes(x, y)) + geom_hex(binwidth=1)
pg <- ggplot_build(p)
pg$data
```

```r
idf <- data.frame(
    x = c(-.5, .5, 0),
    y = c(0, 0, 1 / hh),
    w = c(1, 1, 1),
    h = c(1, 1, 1),
    c = c("a", "b", "c")
)
```

```r
ggplot(idf, aes(x, y)) +
    geom_hex(aes(fill = c, width = w, height = h), stat = "identity") +
    coord_fixed(xlim = c(-1, 1), ylim = c(-hh / 2, 5 * hh / 4))
```

```r
ggplot() +
    geom_hex(aes(fill = "blue", x = 0, y = 0, width = 1 / hh, height = sqrt(3) / 2), stat = "identity") +
    coord_fixed(xlim = c(-.5, .5), ylim = c(-.5, .5))
```

```r
ggplot(idf, aes(x, y)) +
    geom_tile(aes(fill = c, width = w, height = h), stat = "identity")
```

```r
f_hex <- function (df, binwidth=NULL) {
    ggplot(df, aes(x, y)) +
        geom_hex(binwidth = binwidth, color="black") +
        geom_point(color="red", size=4)
}
```

```r
f_rect <- function (df, binwidth=NULL) {
    ggplot(df, aes(x, y)) +
        geom_bin2d(binwidth = binwidth, color="black") +
        geom_point(color="red", size=4)
}
```

```r
f_dat <- function (xlim, ylim) {
    data.frame(
        x = rep(1:xlim, each=ylim),
        y = rep(1:ylim, times=xlim)
    )
}
```

```r
f_hex(f_dat(2, 2))
```

```r
f_rect(f_dat(2, 2))
```

```r
f_hex(f_dat(15, 5))
```

```r
f_rect(f_dat(15, 5))
```

```r
f_hex(f_dat(30, 30))
```

```r
f_rect(f_dat(30, 30))
```

```r
f_hex(f_dat(2, 2), binwidth = c(1, 1))
```

```r
f_rect(f_dat(2, 2), binwidth = c(1, 1))
```

```r
f_hex(f_dat(2, 2), binwidth = c(0.5, 0.5))
```

```r
f_rect(f_dat(2, 2), binwidth = c(0.5, 0.5))
```
