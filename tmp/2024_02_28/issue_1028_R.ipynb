```r
library(ggplot2)
```

```r
df <- data.frame(x = c(0, 1), y = c(0, 1), g = c("A", "B"))
```

```r
p <- ggplot(df, aes(x, y, color = g)) +
    geom_point()
p
```

```r
p + theme(legend.position = 'top')
```

```r
p + theme(legend.position = 'top', legend.justification = 'right')
```

```r
p + theme(legend.position = 'top', legend.justification = c(1, 1))
```

```r
p + theme(legend.position = 'top', legend.justification = c(0, 0))
```
