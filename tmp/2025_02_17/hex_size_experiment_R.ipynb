```r
library(ggplot2)
```

```r
df1 <- data.frame(
    x = c(0, 10),
    y = c(0, 10)
)
```

```r
p11 <- ggplot(df1, aes(x, y)) + geom_hex(stat = 'identity')
p11
```

```r
ggplot_build(p11)$data
```

```r
p12 <- ggplot(df1, aes(x, y)) + geom_hex(aes(width = 10, height = 10 / sqrt(3)), stat = 'identity')
p12
```

```r
ggplot_build(p12)$data
```

```r
df2 <- data.frame(
    x = c(-10, -10, 10, 10),
    y = c(-10, 10, -10, 10)
)
```

```r
p2 <- ggplot(df2, aes(x, y)) + geom_hex(stat = 'identity', color = 'black', linewidth = 1)
p2
```

```r
ggplot_build(p2)$data
```

```r
df3 <- data.frame(
    x = c(-10, -10, 10, 9, 11),
    y = c(-10, 10, -10, 9, 11)
)
```

```r
p31 <- ggplot(df3, aes(x, y)) + geom_hex(color = 'black', size = 1)
p31
```

```r
ggplot_build(p31)$data
```

```r
p32 <- ggplot(df3, aes(x, y)) + geom_hex(binwidth = c(10, 10), color = 'black', size = 1)
p32
```

```r
ggplot_build(p32)$data
```

```r
p33 <- ggplot(df3, aes(x, y)) + geom_hex(aes(width = 1, height = 1), binwidth = c(10, 10), color = 'black', size = 1)
p33
```

```r
ggplot_build(p33)$data
```
