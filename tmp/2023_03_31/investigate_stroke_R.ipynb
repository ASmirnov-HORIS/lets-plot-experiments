# Investigate `stroke`

```r
library(ggplot2)
```

```r
get_plot <- function(shape) {
    ggplot() +
        geom_point(aes(x = 0.0, y = 0.0),
                   size = 50, shape = shape, stroke = 1, color = "red", fill = "black") +
        geom_point(aes(x = 2.0, y = 0.0),
                   size = 50, shape = shape, stroke = 50, color = "red", fill = "black", alpha = 0.5) +
        xlim(-1, 3)
}
```

## 1. Geoms that consumes `stroke`

```r
ggplot() +
    geom_point(aes(x = 0.0, y = 0.0), shape = 21, size = 10, color = "blue", fill = "red") +
    ggtitle("Point: default stroke")
```

```r
ggplot() +
    geom_point(aes(x = 0.0, y = 0.0), shape = 21, size = 10, stroke = 5, color = "blue", fill = "red") +
    ggtitle("Point: stroke = 5")
```

```r
ggplot(iris) +
    geom_boxplot(aes(Species, Sepal.Width),
                 outlier.shape = 21, outlier.size = 10, outlier.stroke = 5, color = "blue", fill = "red") +
    ggtitle("Boxplot: stroke = 5")
```

```r
ggplot() +
    geom_pointrange(aes(x = 0.0, y = 0.5, ymin = 0.0, ymax = 1.0),
                    shape = 21, size = 10, stroke = 5, color = "blue", fill = "red") +
    ggtitle("Point: stroke = 5")
```

```r
ggplot(data.frame(x = c(0, 0, 0, 1)), aes(x)) +
    geom_dotplot(binwidth = .3, color = "blue", fill = "red") +
    ggtitle("Dotplot: default stroke")
```

```r
ggplot(data.frame(x = c(0, 0, 0, 1)), aes(x)) +
    geom_dotplot(binwidth = .3, stroke = 10, color = "blue", fill = "red") +
    ggtitle("Dotplot: stroke = 10")
```

## 2. How `stroke` works

```r
get_plot(21)
```

```r
ggplot() +
    geom_point(aes(x = 0.0, y = 0.0), size = 75, shape = 21, stroke = 75, color = "blue", fill = "red", alpha = 0.25) +
    geom_point(aes(x = 1.0, y = 0.0), size = 75, shape = 21, stroke = 75, color = "blue", fill = "red", alpha = .25) +
    xlim(-2, 3)
```

## 4. Extra

```r
ggplot(data.frame(x = c(0, 0), y = c(0, 0), g = c('a', 'b')), aes(x, y, color = g)) +
    geom_point(shape = 6, size = 20, stroke = 20) +
    facet_grid(cols = vars(g)) +
    scale_x_continuous(limits = c(-2, 2)) +
    coord_flip()
```

```r
ggplot(data.frame(x = c(0, 0, 0, 0), y = c(0, 0, 0, 0), g = c(1, 2, 3, 4)), aes(x, y)) +
    geom_point(aes(size = g, stroke = g), shape = 2, color = "white", position = position_nudge(x = 1)) +
    facet_grid(cols = vars(g)) +
    theme_dark()
```
