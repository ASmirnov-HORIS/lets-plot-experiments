```r
library(ggplot2)
```

```r
df <- data.frame(
    v = c('a', 'b'),
    vmin = c(-2, -1),
    vmax = c(1, 2)
)
```

```r
ggplot(df) +
    geom_errorbar(aes(x = v, ymin = vmin, ymax = vmax), width = .25)
```

```r
ggplot(df) +
    geom_errorbar(aes(y = v, xmin = vmin, xmax = vmax), width = .25)
```
