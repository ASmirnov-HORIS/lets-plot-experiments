```python
import matplotlib.pyplot as plt

def show(value_base, exp_power, *, limits=None, useMathText=None):
    v = value_base * 10**exp_power
    xs = [0, v]
    ys = [0, 0]
    fig, ax = plt.subplots()
    ax.set_title("value: {0}*10^({1})\nlimits: {2}\nMathtext notation: {3}".format(
        value_base, exp_power,
        limits if limits is not None else "default",
        useMathText if useMathText is not None else "default",
    ))
    ax.ticklabel_format(axis='x', scilimits=limits, useMathText=useMathText)
    ax.scatter(xs, ys)
    plt.show()
```

```python
value1 = 1
value2 = -2.5
```

```python
# default limits and useMathText
show(value1, -5)
show(value1, -4)
show(value1, 0)
show(value1, 1)
show(value1, 5)
show(value1, 6)
show(value2, -5)
show(value2, -4)
show(value2, 0)
show(value2, 1)
show(value2, 5)
show(value2, 6)
# default limits and useMathText=True
show(value1, -5, useMathText=True)
show(value1, -4, useMathText=True)
show(value1, 0, useMathText=True)
show(value1, 1, useMathText=True)
show(value1, 5, useMathText=True)
show(value1, 6, useMathText=True)
show(value2, -5, useMathText=True)
show(value2, -4, useMathText=True)
show(value2, 0, useMathText=True)
show(value2, 1, useMathText=True)
show(value2, 5, useMathText=True)
show(value2, 6, useMathText=True)
# custom limits and default useMathText
show(value1, -2, limits=(-2, 3))
show(value1, -1, limits=(-2, 3))
show(value1, 0, limits=(-2, 3))
show(value1, 1, limits=(-2, 3))
show(value1, 2, limits=(-2, 3))
show(value1, 3, limits=(-2, 3))
show(value2, -2, limits=(-2, 3))
show(value2, -1, limits=(-2, 3))
show(value2, 0, limits=(-2, 3))
show(value2, 1, limits=(-2, 3))
show(value2, 2, limits=(-2, 3))
show(value2, 3, limits=(-2, 3))
# custom limits and useMathText=True
show(value1, -2, limits=(-2, 3), useMathText=True)
show(value1, -1, limits=(-2, 3), useMathText=True)
show(value1, 0, limits=(-2, 3), useMathText=True)
show(value1, 1, limits=(-2, 3), useMathText=True)
show(value1, 2, limits=(-2, 3), useMathText=True)
show(value1, 3, limits=(-2, 3), useMathText=True)
show(value2, -2, limits=(-2, 3), useMathText=True)
show(value2, -1, limits=(-2, 3), useMathText=True)
show(value2, 0, limits=(-2, 3), useMathText=True)
show(value2, 1, limits=(-2, 3), useMathText=True)
show(value2, 2, limits=(-2, 3), useMathText=True)
show(value2, 3, limits=(-2, 3), useMathText=True)
# deg 0
show(value1, -1, limits=(0, 0))
show(value1, 0, limits=(0, 0))
show(value1, 1, limits=(0, 0))
show(value2, -1, limits=(0, 0))
show(value2, 0, limits=(0, 0))
show(value2, 1, limits=(0, 0))
```
