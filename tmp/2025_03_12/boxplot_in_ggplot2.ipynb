```r
library(ggplot2)
```

### Default stat

#### First dataset

```r
df11 <- data.frame(
    cat = c('a', 'a', 'a', 'a', 'b', 'b', 'b', 'b'),
    val = c(-3, -2, 0, 1, -1, 0, 2, 3)
)
```

```r
# Default orientation
ggplot() + geom_boxplot(aes(cat, val), data = df11)
```

```r
# Swap aesthetics
ggplot() + geom_boxplot(aes(val, cat), data = df11)
```

```r
# Default aesthetics and orientation = 'y'
ggplot() + geom_boxplot(aes(cat, val), data = df11, orientation = 'y')
```

```r
# Swap aesthetics and orientation = 'y'
ggplot() + geom_boxplot(aes(val, cat), data = df11, orientation = 'y')
```

### `stat = 'identity'`

#### First dataset

```r
df21 <- data.frame(
    cat = c('a', 'b'),
    min = c(-3, -1),
    lower = c(-2, 0),
    middle = c(-1, 1),
    upper = c(0, 2),
    max = c(1, 3)
)
```

```r
# Default orientation
ggplot() + geom_boxplot(aes(x = cat, ymin = min, lower = lower, middle = middle, upper = upper, ymax = max), data = df21, stat = 'identity')
```

```r
# Swap aesthetics
ggplot() + geom_boxplot(aes(y = cat, xmin = min, xlower = lower, xmiddle = middle, xupper = upper, xmax = max), data = df21, stat = 'identity')
```

```r
# Default aesthetics and orientation = 'y'
ggplot() + geom_boxplot(aes(x = cat, ymin = min, lower = lower, middle = middle, upper = upper, ymax = max), data = df21, stat = 'identity', orientation = 'y')
```

```r
# Swap aesthetics and orientation = 'y'
ggplot() + geom_boxplot(aes(y = cat, xmin = min, xlower = lower, xmiddle = middle, xupper = upper, xmax = max), data = df21, stat = 'identity', orientation = 'y')
```
