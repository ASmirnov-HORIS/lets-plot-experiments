```r
library(ggplot2)
```

```r
df <- data.frame(
    x = c(-1, -1, 1, 1, 1),
    y = c(-1, 1, -1, 1, 1)
)
```

```r
ggplot(df, aes(x, y)) + geom_hex(bins=2)
```

```r
ggplot(df, aes(x, y)) + geom_hex(binwidth=1)
```

```r
p <- ggplot(df, aes(x, y)) + geom_hex(binwidth=1)
pg <- ggplot_build(p)
pg$data
```

```r
idf <- data.frame(
    x = c(-.5, .5, 0),
    y = c(0, 0, 2),
    w = c(1, 1, 1),
    h = c(1, 1, 1),
    c = c("a", "b", "c")
)
```

```r
ggplot(idf, aes(x, y)) +
    geom_hex(aes(fill = c, width = w, height = h), stat = "identity")
```

```r
ggplot() +
    geom_hex(aes(fill = "blue", x = 0, y = 0, width = 1, height = sqrt(3) / 2), stat = "identity") +
    coord_fixed(xlim = c(-.5, .5), ylim = c(-.5, .5))
```

```r
ggplot(idf, aes(x, y)) +
    geom_tile(aes(fill = c, width = w, height = h), stat = "identity")
```

```r
df <- data.frame(
    x = c(0, 0, 0, 2.5, 2.5, 2.5, 5, 5, 5),
    y = c(0, 3.5, 5, 0, 3.5, 5, 0, 3.5, 5)
)

ggplot(df, aes(x, y)) +
    geom_hex(binwidth=10) +
    geom_point(color="red", size=4) +
    coord_fixed(xlim=c(-5, 5), ylim=c(-5, 5))
```

```r
hc = 2 / sqrt(3)
```

```r
f <- function (df) {
    ggplot(df, aes(x, y)) +
        geom_hex(binwidth=2) +
        geom_point(color="red", size=4) +
        coord_fixed(xlim=c(-1, 1), ylim=c(-1.5, 1.5))
}
```

```r
f(data.frame(
    x = c(0, 0),
    y = c(0, hc)
))
```

```r
f(data.frame(
    x = c(0, 0.5),
    y = c(0, 3 * hc / 4)
))
```

```r
f(data.frame(
    x = c(0, 1),
    y = c(0, hc / 2)
))
```

```r
f(data.frame(
    x = c(0, 1),
    y = c(0, 0)
))
```

```r
f(data.frame(
    x = c(0, 1),
    y = c(0, -hc / 2)
))
```

```r
f(data.frame(
    x = c(0, 0.5),
    y = c(0, -3 * hc / 4)
))
```

```r
f(data.frame(
    x = c(0, 0),
    y = c(0, -hc)
))
```

```r
f(data.frame(
    x = c(0, -0.5),
    y = c(0, -3 * hc / 4)
))
```

```r
f(data.frame(
    x = c(0, -1),
    y = c(0, -hc / 2)
))
```
