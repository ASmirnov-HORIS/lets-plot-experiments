```r
library(ggplot2)
```

```r
pointrange <- function(x, fatten, size, stroke, linewidth) {
    geom_pointrange(aes(x = x, y = 1, ymin = 0, ymax = 1),
                    shape = 21, color = "blue", fill = "red", alpha = 0.5,
                    fatten = fatten, size = size, stroke = stroke, linewidth = linewidth)
}
```

```r
ggplot() +
    pointrange(1, 1, 2, 1, 1) +
    pointrange(2, 1, 2, 1, 4) +
    pointrange(3, 1, 2, 4, 1) +
    pointrange(4, 1, 2, 4, 4) +
    pointrange(5, 1, 5, 1, 1) +
    pointrange(6, 1, 5, 1, 4) +
    pointrange(7, 1, 5, 4, 1) +
    pointrange(8, 1, 5, 4, 4) +
    pointrange(9, 4, 2, 1, 1) +
    pointrange(10, 4, 2, 1, 4) +
    pointrange(11, 4, 2, 4, 1) +
    pointrange(12, 4, 2, 4, 4) +
    pointrange(13, 4, 5, 1, 1) +
    pointrange(14, 4, 5, 1, 4) +
    pointrange(15, 4, 5, 4, 1) +
    pointrange(16, 4, 5, 4, 4)
```
