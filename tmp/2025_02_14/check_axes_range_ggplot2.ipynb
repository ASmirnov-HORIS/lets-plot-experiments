```r
library(ggplot2)
```

### Point

```r
ggplot() + geom_point(aes(x = c(0, 1), y = 0), size = 100)
```

### Hexagon

```r
ggplot() + geom_hex(aes(x = c(0, 1), y = 0, width = .1, height = .01), stat = 'identity')
```
