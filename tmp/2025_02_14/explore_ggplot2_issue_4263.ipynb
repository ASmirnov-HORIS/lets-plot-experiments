```r
library(ggplot2)
```

```r
d <- data.frame(y=rep(1:5, times=5),
                x=rep(1:5, each=5) + c(0.5, 0, 0.5, 0, 0.5),
                m='g')
d[7, 3] <- 'b'
head(d, 8)
```

```r
g <- ggplot() +
    geom_hex(data=d,
             aes(x=x, y=y, fill=m),
             stat='identity') 
plot(g)
```

```r
d <- data.frame(y=rep(1:5, times=5),
                x=rep(1:5, each=5) + c(0.5, 0, 0.5, 0, 0.5),
                m='g')
d[7, 3] <- 'X'
head(d, 8)
```

```r
g <- ggplot() +
    geom_hex(data=d,
             aes(x=x, y=y, fill=m),
             stat='identity') 
plot(g)
```

```r
d <- data.frame(
  y = rep(1:5, times = 5),
  x = rep(1:5, each = 5) + c(0.5, 0, 0.5, 0, 0.5),
  m = "g"
)
d[7, 3] <- "b"
head(d, 8)
```

```r
p <- ggplot(d, aes(x = x, y = y, fill = m)) +
  geom_point() +
  geom_hex(alpha = 0.3, stat = "identity")

p
```

```r
p + facet_wrap(~m)
```

```r
ggplot(d, aes(x = x, y = y, fill = m, group = 1)) +
  geom_point() +
  geom_hex(alpha = 0.3, stat = "identity")
```
