```r
library(ggplot2)
```

```r
df <- data.frame(
    x = c(0, 0, 2, 0, 0, 8),
    y = c(0, 1, 0, 0, 2, 0),
    f = c('a', 'a', 'a', 'b', 'b', 'b')
)
```

```r
p <- ggplot(df, aes(x, y)) + geom_point()
p
```

```r
fixed_p <- p + coord_fixed()
fixed_p
```

```r
fixed_p + facet_grid(cols = vars(f))
```

```r
fixed_p + facet_grid(cols = vars(f), scales = "free")
```
