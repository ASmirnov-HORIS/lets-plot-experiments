# Legend Justification

```r
library(ggplot2)
library(repr)
library(grid)
```

```r
theme_default <- theme(
    plot.background = element_rect(colour = "green", linewidth=3),
    panel.background = element_rect(colour = "red", linewidth=3),
    legend.box.background = element_rect(colour = "blue", linewidth=3)
)
```

```r
legend_pos1_05_just_0_05 <- theme(legend.position = c(1, 0.5), legend.justification = c(0, 0.5))
```

```r
legend_pos95_95_just_r_t <- theme(legend.position = c(.95, .95), legend.justification = c("right", "top"))
```

```r
df <- data.frame(
    x = c(-1, 1, -1, 1),
    y = c(-1, -1, 1, 1),
    z = c(1, 2, 3, 4)
)
```

```r
p <- ggplot(df, aes(x, y)) + geom_tile(aes(fill = z))
```

### Default

```r
p1 <- p + theme_default
p1
```

```r
p2 <- p1 + legend_pos1_05_just_0_05
p2
```

```r
p3 <- p1 + legend_pos95_95_just_r_t
p3
```

### `ggsave()`

```r
ggsave("v1.png", plot = p1, width = 1000, height = 500,  units = "px")
```

```r
ggsave("v2.png", plot = p2, width = 1000, height = 500,  units = "px")
```

```r
ggsave("v3.png", plot = p3, width = 1000, height = 500,  units = "px")
```
