```r
library(ggplot2)
library(ggforce)
```

```r
head(mtcars)
```

```r
number_to_word <- c(`3` = "three", `4` = "four", `5` = "five")

mtcars$gear <- number_to_word[as.character(mtcars$gear)]

mtcars$gear <- as.character(mtcars$gear)

head(mtcars)
```

```r
# graph 1
p1 = ggplot(mtcars, aes(x=as.factor(cyl), y=hp)) + geom_violin(bw=20) + geom_sina()
p1
```

```r
# graph 2
p2 = ggplot(mtcars, aes(x=as.factor(cyl), y=hp, color=gear)) + geom_violin(bw=20, drop=FALSE) + geom_sina()
p2
```

```r
# graph 3
p3 = ggplot(mtcars, aes(x=as.factor(cyl), y=hp, color=gear, group=cyl)) + geom_violin(bw=20, drop=FALSE) + geom_sina()
p3
```
