```python
import pandas as pd

from lets_plot import *
```

```python
LetsPlot.setup_html()
LetsPlot.set_theme(theme_gray())
```

```python
df = pd.read_csv("data/usgs_earthquakes_2025_m45.csv", parse_dates=["time"])
print(df.shape)
df.head()
```

```python
# histogram breaks

breaks = [4.5, 4.6, 4.8, 5, 5.5, 9]

ggplot(df, aes(x="mag")) + \
    geom_histogram(breaks=breaks) + \
    ggtitle("Magnitude distribution")
```

```python
# pointdensity

p = ggplot(df.sort_values(by="mag"), aes("longitude", "latitude")) + \
    geom_livemap() + \
    theme(legend_position='bottom')
    
gggrid([
    p + \
        geom_point(aes(color="mag")) + \
        scale_color_viridis(option='magma') + \
        ggtitle("Epicenters: magnitude"),
    p + \
        geom_pointdensity() + \
        scale_color_viridis() + \
        ggtitle("Epicenters: density")
], ncol=1) + ggsize(800, 1000)
```

```python
# legend wraps automatically

ggplot(df, aes("mag", "depth", color="magType")) + \
    geom_point(alpha=.5) + \
    theme(legend_position='bottom') + \
    ggtitle("Depth vs. Magnitude")
```

```python
# facet_wrap(drop=...)
# theme(panel_spacing=..., strip_spacing=...)

ggplot(df.assign(depth_big=(df["depth"] >= 350).map({True: "depth >= 350", False: "depth < 350"}))) + \
    geom_point(aes("mag", "depth", color="magType"), alpha=.5) + \
    facet_wrap(["magType", "depth_big"], drop=False) + \
    theme(panel_spacing=4, strip_spacing=2)
```

```python
ggplot(df.sort_values(by="depth"), aes(sample="mag")) + \
    geom_qq(aes(color="depth"), distribution='exp', alpha=.5,
            tooltips=layer_tooltips().line("@|@magType").line("@|@mag").line("@|@depth")) + \
    ggtitle("Q-Q plot of magnitudes", "distribution='exp'")
```
