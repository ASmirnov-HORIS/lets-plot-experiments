```r
library(ggplot2)
```

```r
get_data <- function(n) {
  x <- 0:(n-1)
  a <- sin(x / 10.0) * 20.0
  b <- sin(x / 6.0 + 1.0) * 20.0 - 2.0
  
  data.frame(
    x = x,
    min = pmin(a, b),
    max = pmax(a, b),
    label = ifelse(a > b, "a", "b")
  )
}

data <- get_data(60)
```

```r
ggplot(data) +
    geom_ribbon(aes(x = x, ymin = min, ymax = max, fill = label, color = label), alpha = 0.2)
```
