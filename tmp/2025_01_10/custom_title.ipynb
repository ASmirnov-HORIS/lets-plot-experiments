## Title Customization

```python
from lets_plot import *
from lets_plot.bistro import *
```

```python
LetsPlot.setup_html()
LetsPlot.set_theme(theme(plot_title=element_text(face='bold')))
```

```python
title = ggtitle("Title")
```

## Default

```python
ggplot() + geom_point(x=0) + title
```

## Theme and Flavor

```python
ggplot() + geom_point(x=0) + theme_bw() + title
```

```python
ggplot() + geom_point(x=0) + flavor_solarized_light() + title
```

## Bistro

```python
corr_plot({'x': [1, 2, 3], 'y': [1, 1, 2]}).tiles().build() + title
```

```python
qq_plot({'x': [1, 2, 2]}, sample='x') + title
```

```python
joint_plot({'x': [1, 2, 3], 'y': [1, 2, 2]}, 'x', 'y') + title
```

```python
residual_plot({'x': [1, 2, 3], 'y': [1, 2, 2]}, 'x', 'y') + title
```

```python
waterfall_plot({'x': ['A'], 'y': [1]}, 'x', 'y') + title
```
