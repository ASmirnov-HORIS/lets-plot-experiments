```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
def plot(**override_aes):
    return ggplot({'x': [1, 2, 3, 3], 'g': ['a', 'a', 'b', 'b']}) + \
        geom_bar(aes(x='x', fill='g'), color='black') + \
        scale_fill_brewer(guide=guide_legend(override_aes=override_aes))

gggrid([
    plot() + ggtitle("Default"),
    plot(size=10) + ggtitle("size=10"),
    plot(stroke=10) + ggtitle("stroke=10"),
    plot(linewidth=10) + ggtitle("linewidth=10"),
], ncol=2)
```
