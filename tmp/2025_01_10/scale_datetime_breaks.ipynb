```python
from datetime import datetime

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
def get_dt(year=2024, month=1, day=1):
    return int(round(1_000 * datetime(year, month, day).timestamp()))
```

```python
xs = [get_dt(month=m, day=d) for m in range(1, 13) for d in range(1, 29)]
ys = [int(v / 10**5) % 29 for v in xs]
data = {'x': xs, 'y': ys}
```

```python
p = ggplot(data, aes('x', 'y')) + geom_line()
```

```python
p + scale_x_datetime(breaks=[get_dt(month=month, day=1) for month in range(1, 13)], format="%b")
```

```python
p + scale_x_datetime(breaks=[get_dt(month=month, day=2) for month in range(1, 13)], format="%b")
```
