```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
def get_plot(data):
    return ggplot(data) + geom_density2d(aes('x', 'y', weight='w')) + geom_point(aes('x', 'y', color='w'), size=7)
```

```python
data1 = {
    'x': [0, 0, 1, 1],
    'y': [0, 1, 0, 1],
    'w': [1, 1, 2, 1],
}
data2 = {
    'x': [0, 0, 0.5, 1, 1],
    'y': [0, 1, None, 0, 1],
    'w': [1, 1, 1, 2, 1],
}

gggrid([
    get_plot(data1) + ggtitle("Expected"),
    get_plot(data2) + ggtitle("Actual (weights are shifted)"),
])
```
