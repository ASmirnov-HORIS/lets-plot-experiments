```r
library(ggplot2)
```

```r
data <- data.frame(
    x = c("A", "A", "B", "B", "C", "C"),
    ymin = c(0, 1, -1, 0, -1, 1),
    y = c(2, 2, 1, 2, 3, 3),
    ymax = c(3, 4, 3, 3, 5, 4),
    g = c("g1", "g2", "g1", "g2", "g1", "g2")
)
```

```r
set.seed(42)
n <- 100
df <- data.frame(
    x = sample(c("A", "B", "C"), n, replace=TRUE),
    y = rnorm(n, 0, 1),
    u = runif(n, 0, 1),
    g = sample(c("g1", "g2"), n, replace=TRUE)
)
```

```r
ggplot(data) + geom_errorbar(aes(x, y, ymin = ymin, ymax = ymax, color = g), linewidth = 1)
```

```r
ggplot(data) + geom_crossbar(aes(x, y, ymin = ymin, ymax = ymax, fill = g), linewidth = 1)
```

```r
ggplot(data) + geom_linerange(aes(x, ymin = ymin, ymax = ymax, color = g), linewidth = 1)
```

```r
ggplot(data) + geom_pointrange(aes(x, y, ymin = ymin, ymax = ymax, color = g), linewidth = 1)
```

```r
ggplot(df) + geom_boxplot(aes(x, y, fill = g), linewidth = 1)
```

```r
ggplot(df) + geom_violin(aes(x, y, fill = g), linewidth = 1)
```

```r
ggplot(df) + stat_summary(aes(x, y, color = g), fun.data = "mean_se", linewidth = 1)
```

```r
ggplot(df) + stat_summary_bin(aes(u, y, color = g), fun.data = "mean_se", binwidth = .2, linewidth = 1)
```
