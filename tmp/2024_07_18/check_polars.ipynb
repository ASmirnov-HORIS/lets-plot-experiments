```python
import pandas as pd
import polars as pl

from lets_plot import *
from lets_plot.bistro import *
```

```python
LetsPlot.setup_html()
```

```python
def get_plot(plot_builder):
    def build(wrap_data):
        wrap_data_str = str(wrap_data).replace("<class '", "").replace("'>", "")
        try:
            (plot_builder(wrap_data) + ggtitle(wrap_data_str) + ggsize(400, 300)).show()
            return
        except:
            print("Problem with {0}".format(wrap_data_str))
            return
    build(dict)
    build(pd.DataFrame)
    build(pl.DataFrame)
```

```python
def get_corr_plot(wrap_data):
    data = {'x': [0, 0, 1], 'y': [0, 1, 1]}
    return corr_plot(wrap_data(data)).tiles().build()

get_plot(get_corr_plot)
```

```python
def get_qq_plot(wrap_data):
    data = {'s': [0, 0, 1]}
    return qq_plot(wrap_data(data), sample='s')

get_plot(get_qq_plot)
```

```python
def get_joint_plot(wrap_data):
    data = {'x': [0, 0, 1], 'y': [0, 1, 1]}
    return joint_plot(wrap_data(data), 'x', 'y')

get_plot(get_joint_plot)
```

```python
def get_residual_plot(wrap_data):
    data = {'x': [0, 0, 1], 'y': [0, 1, 1]}
    return residual_plot(wrap_data(data), 'x', 'y')

get_plot(get_residual_plot)
```
