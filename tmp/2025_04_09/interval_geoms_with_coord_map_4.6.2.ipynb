```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
data1 = {
    'cat': [-15, -15, -15, -16, -16, -16],
    'val': [28, 29, 31, 29, 31, 32],
}

gggrid([
    ggplot(data1) + geom_boxplot(aes('cat', 'val')) + ggtitle("Default coordinates"),
    ggplot(data1) + geom_boxplot(aes('cat', 'val')) + coord_map() + ggtitle("coord_map()"),
    ggplot(data1) + geom_boxplot(aes('val', 'cat'), orientation='y') + ggtitle("Default coordinates"),
    ggplot(data1) + geom_boxplot(aes('val', 'cat'), orientation='y') + coord_map() + ggtitle("coord_map()"),
], ncol=2)
```

```python
data2 = {
    'cat': [-15, -16],
    'min': [29, 30],
    'max': [30, 31],
}

gggrid([
    ggplot(data2) + geom_errorbar(aes(x='cat', ymin='min', ymax='max')) + ggtitle("Default coordinates"),
    ggplot(data2) + geom_errorbar(aes(x='cat', ymin='min', ymax='max')) + coord_map() + ggtitle("coord_map()"),
    ggplot(data2) + geom_errorbar(aes(y='cat', xmin='min', xmax='max')) + ggtitle("Default coordinates"),
    ggplot(data2) + geom_errorbar(aes(y='cat', xmin='min', xmax='max')) + coord_map() + ggtitle("coord_map()"),
], ncol=2)
```
