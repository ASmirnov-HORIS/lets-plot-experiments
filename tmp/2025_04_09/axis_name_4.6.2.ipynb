```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
data = {
    'cat': ['a', 'a', 'a', 'b', 'b', 'b'],
    'val': [20, 30, 35, 25, 30, 40],
}

gggrid([
    ggplot(data) + geom_boxplot(aes('cat', 'val')) + ggtitle("boxplot geom + boxplot stat"),
    ggplot(data) + stat_summary(aes('cat', 'val')) + ggtitle("pointrange geom + summary stat"),
    ggplot(data) + geom_errorbar(aes('cat', 'val'), stat='summary') + ggtitle("pointrange geom + summary stat"),
    ggplot(data) + geom_boxplot(aes('cat', 'val', lower='..ymin..', upper='..ymax..'), stat='summary') + ggtitle("boxplot geom + summary stat"),
    ggplot(data) + geom_ribbon(aes('cat', 'val'), stat='summary') + ggtitle("ribbon geom + summary stat")
], ncol=2)
```
