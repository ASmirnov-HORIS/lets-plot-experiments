```python
import itertools
import pandas as pd
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
df = pd.DataFrame({
    'x': [0, 0, 1, 1],
    'y': [0, 1, 0, 1],
    'z': [0, 1, 2, 3],
})

ggplot(df) + geom_contour(aes('x', 'y', z='z')) + coord_fixed()
```

```python
ggplot(df.sort_values(by=["y", "x"])) + geom_contour(aes('x', 'y', z='z')) + coord_fixed()
```
