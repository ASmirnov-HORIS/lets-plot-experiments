```r
library(ggplot2)
```

```r
df <- data.frame(
    x1 = c(-10, -1, 0, 1, 10, 100),
    y1 = c(-3, -2, -1, 1, 2, 3),
    x2 = c(-100, -10, -1, 0, 1, 10),
    y2 = c(-3, -2, -1, 1, 2, 3)
)
```

```r
p1 <- ggplot(df) +
    geom_segment(aes(x1, y1, xend = x2, yend = y2), linewidth = 1) +
    geom_point(aes(x1, y1), size = 3) +
    geom_point(aes(x2, y2), size = 3)
p1 + ggtitle("Default")
```

```r
p1 + scale_x_log10() + ggtitle("scale_x_log10()")
```

```r
ggplot(tail(df, 3)) +
    geom_segment(aes(x1, y1, xend = x2, yend = y2), linewidth = 1) +
    geom_point(aes(x1, y1), size = 3) +
    geom_point(aes(x2, y2), size = 3) +
    scale_x_log10() +
    ggtitle("Only non-negative values")
```
