```python
from lets_plot import *
LetsPlot.setup_html()
```

```python
# Data and colors
xs = ['a', 'b', 'c', 'd', 'e', 'f']
data = {'x': xs}
colors = ["red", "green", "blue", "yellow", "orange", "pink"]
```

```python
# Base plot
p = (ggplot(data) +
     geom_tile(aes(x='x', fill=colors)) +
     scale_fill_identity(guide=guide_legend(ncol=2)) +
     ggsize(600, 200))
p
```

```python
plot = (ggplot(data) +
        ggtitle("Overriding aes.") +
        geom_tile(aes(x='x', fill=colors)) +
        guides(alpha='none', fill=guide_legend(nrow=2, override_aes={'color': 'black'})) +
        ggsize(600, 200))

plot.show()
```

```python
plot = (ggplot(data) +
        ggtitle("Overriding aes.") +
        geom_tile(aes(x='x', fill=colors)) +
        scale_fill_identity(guide=guide_legend()) +
        guides(alpha='none', fill=guide_legend(nrow=2, override_aes={'color': 'black'})) +
        ggsize(600, 200))

plot.show()
```
