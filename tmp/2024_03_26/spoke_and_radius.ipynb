```r
library(ggplot2)
```

```r
df <- data.frame(
    x = c(0, 0, 1, 1),
    y = c(0, 1, 0, 1),
    a = c(5*pi/4, 3*pi/4, 7*pi/4, pi/4),
    r = c(.75, .5, 1, .25)
)
```

### Basic example

```r
p1 <- ggplot(df) +
    geom_spoke(aes(x, y, angle = a, radius = r))
p1
```

### With `scale_radius()`

```r
my_scale <- scale_radius(name = "radius", range = c(1, 10))
```

```r
p1 + my_scale
```

### It works with points

```r
p2 <- ggplot(df) +
    geom_point(aes(x, y, size = r))
p2
```

```r
p2 + my_scale
```
