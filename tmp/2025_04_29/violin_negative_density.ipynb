```python
LetsPlot.setup_html()
```

```python
data = {
    'cat': [5] * 3 + [10] * 3,
    'val': [1, 2, 2, 1, 1, 1],
    'weight': [-2] * 6,
}
```

```python
g = ggplot(data, aes('cat', 'val'))
```

```python
g + geom_violin(show_half=1, scale='count')
```

```python
g + geom_violin(aes(weight='weight'), show_half=1, scale='count')
```
