```python
import numpy as np

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
data = {
    'val': [1, 2, 2, 1, 1, 2],
    'group': ['g1'] * 3 + ['g2'] * 3,
}
```

```python
g1 = ggplot(data, aes(y='val', color='group'))
```

```python
gggrid([
    g1 + geom_crossbar(stat='boxplot'),
    g1 + geom_crossbar(stat='boxplot', position=position_dodge(width=.95)),
])
```

```python
gggrid([
    g1 + geom_boxplot(),
    g1 + geom_boxplot(position=position_dodge(width=.95)),
])
```

```python
gggrid([
    g1 + geom_violin(),
    g1 + geom_violin(position=position_dodge(width=.95)),
])
```

```python
jitter_data = {'x': [0] * 10_000, 'y': [0] * 10_000}
g2 = ggplot(jitter_data, aes('x', 'y')) + coord_fixed(xlim=[-1, 1], ylim=[-1, 1])

gggrid([
    g2 + geom_jitter(seed=42),
    g2 + geom_jitter(seed=42, position=position_jitter(width=.4, height=.4)),
])
```

```python
nudge_data = {'x': [0], 'y': [0]}
g3 = ggplot(nudge_data, aes('x', 'y')) + coord_fixed(xlim=[-1, 1], ylim=[-1, 1])

gggrid([
    g3 + geom_point(position=position_nudge()),
    g3 + geom_point(position=position_nudge(x=.1, y=.1)),
])
```

```python
jitter_data = {'x': ['a'] * 10_000, 'y': [0] * 10_000}
g4 = ggplot(jitter_data, aes('x', 'y')) + coord_fixed(xlim=[-1, 1], ylim=[-1, 1])

gggrid([
    g4 + geom_point(position=position_jitterdodge(seed=42)),
    g4 + geom_point(position=position_jitterdodge(seed=42, jitter_width=.4, jitter_height=.4)),
])
```

```python
stack_data = {'x': [0] * 3, 'y': [2] * 3}
g5 = ggplot(stack_data, aes('x', 'y')) + coord_fixed(xlim=[-1, 1], ylim=[0, 10])

gggrid([
    g5 + geom_point(position=position_stack(mode='all')),
    g5 + geom_point(position=position_stack(mode='all', vjust=1)),
])
```

```python
gggrid([
    ggplot(stack_data, aes('x', 'y')) + geom_point(position=position_fill(mode='all')),
    ggplot(stack_data, aes('x', 'y')) + geom_point(position=position_fill(mode='all', vjust=1)),
])
```

```python
data = {
    'x': ['a'] * 3 + ['b'] * 3,
    'y': [1, 1, 2, 1, 2, 2],
}
```

```python
g = ggplot(data, aes('x', 'y'))
```

```python
gggrid([
    g + geom_violin(fill="red") + geom_violin(width=.5, fill="blue"),
])
```

```python
n = 25
tvalues = [i / n for i in range(n + 1)]
p = ggplot(data, aes('x', 'y'))
for t in tvalues:
    if t == 0:
        size = 1
    else:
        size = 0
    p += geom_violin(size=size, fill="blue", alpha=t**2, width=1*(1-t))
p
```
